```python
import os
import shutil
import holidays
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from google.colab import drive

# Library Statistik untuk EDA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Preprocessing & Metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error

# Deep Learning (TensorFlow)
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Bidirectional, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Tuning
import keras_tuner as kt

print("Library berhasil diimpor.")
```

# Data Preparation

## Load Data

```python
drive.mount('/content/drive')

base_path = '/content/drive/MyDrive/Magang Prima PTKI/data_dummy'
```

```python
# Load Data
transactions_df = pd.read_csv(os.path.join(base_path, 'transactions.csv'))
transaction_details_df = pd.read_csv(os.path.join(base_path, 'transaction_details.csv'))
products_df = pd.read_csv(os.path.join(base_path, 'products.csv'))
stores_df = pd.read_csv(os.path.join(base_path, 'stores.csv'))
```

## Merging Data

```python
# Merging
df_merged = pd.merge(transactions_df, transaction_details_df, left_on='id', right_on='transactionId')
df_merged = pd.merge(df_merged, products_df[['id', 'productCategoryId']], left_on='productId', right_on='id', suffixes=('', '_prod'))
df_merged = pd.merge(df_merged, stores_df[['id', 'storeTypeId']], left_on='storeId', right_on='id', suffixes=('', '_store'))
```

```python
# Filter Store ID 1 (Dealer Otomotif Sejahtera)
TARGET_STORE_ID = 1
df_automotive = df_merged[df_merged['storeId'] == TARGET_STORE_ID].copy()
df_automotive['transactionDate'] = pd.to_datetime(df_automotive['transactionDate'])
```

## Data Cleaning

```python
# Agregasi Harian
time_series_automotive = df_automotive.groupby(df_automotive['transactionDate'].dt.date)['subtotal'].sum().reset_index()
time_series_automotive.columns = ['date', 'revenue']
time_series_automotive['date'] = pd.to_datetime(time_series_automotive['date'])
time_series_automotive.set_index('date', inplace=True)
```

```python
# Reindex agar tanggal berurutan
all_days = pd.date_range(start=time_series_automotive.index.min(), end=time_series_automotive.index.max(), freq='D')
time_series_automotive = time_series_automotive.reindex(all_days, fill_value=0)
```

```python
# Handling Outliers (Clipping)
Q1 = time_series_automotive['revenue'].quantile(0.25)
Q3 = time_series_automotive['revenue'].quantile(0.75)
IQR = Q3 - Q1
time_series_automotive['revenue_clean'] = time_series_automotive['revenue'].clip(lower=Q1 - 1.5*IQR, upper=Q3 + 1.5*IQR)

print(f"Data Otomotif (Store {TARGET_STORE_ID}) berhasil disiapkan.")
```

# Preprocessing Data

## Feature Engineering

(Menambah Konteks Waktu)

```python
# 1. Feature Engineering
id_holidays = holidays.Indonesia(years=[2025, 2026])
time_series_automotive['is_holiday'] = time_series_automotive.index.strftime('%Y-%m-%d').isin(id_holidays).astype(int)
time_series_automotive['is_payday'] = time_series_automotive.index.day.map(lambda x: 1 if x >= 25 or x <= 5 else 0)
time_series_automotive['is_weekend'] = time_series_automotive.index.weekday.map(lambda x: 1 if x >= 5 else 0)
```

```python
# Fitur Lag 1-7
for i in range(1, 8):
    time_series_automotive[f'lag_{i}'] = time_series_automotive['revenue_clean'].shift(i)
```

```python
# Fitur Tren (Rolling Average 7 hari)
time_series_automotive['rolling_avg_7d'] = time_series_automotive['revenue_clean'].rolling(window=7).mean()
```

```python
# Update Feature Engineering (Menambah Day of Month)
time_series_automotive['day_of_month'] = time_series_automotive.index.day
```

```python
# Hapus baris yang kosong akibat proses Lag
time_series_automotive.dropna(inplace=True)
```

## Scaling

```python
# 2. Definisikan X dan y
features = [f'lag_{i}' for i in range(1, 8)] + ['is_holiday', 'is_payday', 'is_weekend', 'rolling_avg_7d', 'day_of_month']
X_raw = time_series_automotive[features]
y_raw = time_series_automotive['revenue_clean']
```

```python
# 3. Scaling (Mengubah angka jutaan ke skala 0-1)
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
```

```python
X_scaled = scaler_X.fit_transform(X_raw)
y_scaled = scaler_y.fit_transform(y_raw.values.reshape(-1, 1))
```

```python
# 4. Reshaping ke format 3D untuk LSTM [Samples, Time Steps, Features]
X_final = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))
y_final = y_scaled
```

## Splitting Data (Train & Test)

```python
# 5. Split Data (80% Training, 20% Testing)
train_size = int(len(X_final) * 0.8)
X_train, X_test = X_final[:train_size], X_final[train_size:]
y_train, y_test = y_scaled[:train_size], y_scaled[train_size:]

print(f"Preprocessing Selesai! Data siap dilatih dengan {X_train.shape[2]} fitur.")
```

# Modeling

```python
def model_builder(hp):
    model = Sequential()

    # CNN Layer
    model.add(Conv1D(
        filters=hp.Int("conv_filters", 32, 128, 16),
        kernel_size=1, activation='relu',
        input_shape=(X_train.shape[1], X_train.shape[2])
    ))
    model.add(MaxPooling1D(pool_size=1))

    # Bi-LSTM Layer
    model.add(Bidirectional(LSTM(units=hp.Int("bilstm_units", 64, 256, 32), return_sequences=True)))
    model.add(Dropout(hp.Float("dropout_1", 0.1, 0.4, 0.1)))

    # LSTM 2
    model.add(LSTM(units=hp.Int("lstm_units", 32, 128, 32), return_sequences=False))

    model.add(Dense(hp.Int("dense_units", 32, 128, 32), activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
    return model

# Reset Tuner Khusus
tuner_path = 'tuner_dir/automotive_tuning'
if os.path.exists(tuner_path):
    shutil.rmtree(tuner_path)

tuner = kt.RandomSearch(
    model_builder,
    objective='val_loss',
    max_trials=10,
    directory='tuner_dir',
    project_name='automotive_tuning',
    overwrite=True
)

early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

print("Memulai Hyperparameter Tuning untuk Otomotif...")
tuner.search(X_train, y_train, epochs=50, validation_split=0.2, callbacks=[early_stop], verbose=1)

# Ambil Model Terbaik
best_model = tuner.get_best_models(num_models=1)[0]
print("\nTuning Selesai. Arsitektur model terbaik dipilih.")
best_model.summary()
```

# Evaluation & Visualization

```python
# 1. Pelatihan Akhir Model Terbaik
print("Melatih model akhir...")
history = best_model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# 2. Prediksi & Kembalikan Skala (Inverse Transform)
y_pred_scaled = best_model.predict(X_test)
```

```python
# 3. Inverse Scale (Kembalikan ke nilai Rupiah asli)
y_pred = scaler_y.inverse_transform(y_pred_scaled).ravel()
y_actual = scaler_y.inverse_transform(y_test).ravel()
```

```python
# 4. Hitung Metrik Utama
mae = mean_absolute_error(y_actual, y_pred)
mape = mean_absolute_percentage_error(y_actual, y_pred) * 100
r2 = r2_score(y_actual, y_pred)
```

```python
print("="*50)
print("     HASILEVALUASI MODEL OTOMOTIF (STORE 1)")
print("="*50)
print(f"1. Mean Absolute Error (MAE)  : Rp {mae:,.2f}")
print(f"2. MAPE (Tingkat Error)       : {mape:.2f}%")
print(f"3. R2-Score (Akurasi Model)   : {r2:.4f}")
print("="*50)
```

```python
# 4. Visualisasi 1: Prediksi vs Aktual
plt.figure(figsize=(15, 6))
plt.plot(y_actual, label='Data Aktual (Automotive)', color='black', linewidth=2)
plt.plot(y_pred, label='Prediksi AI (Automotive)', color='slategray', linestyle='--')
plt.title(f'Akurasi Prediksi Otomotif (MAPE: {mape:.2f}% | R2: {r2:.4f})')
plt.ylabel('Pendapatan (IDR)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 5. Visualisasi 2: Loss Curve
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss', color='blue')
plt.plot(history.history['val_loss'], label='Val Loss', color='red')
plt.title('Grafik Training vs Validation Loss (Otomotif)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Kode Inference Dengan Analisis Bisnis

```python
# 1. Load Model & Scaler
model = load_model('/content/drive/MyDrive/Magang Prima PTKI/models/otomotif/model_automotive_store_1.keras')
scaler_X = joblib.load('/content/drive/MyDrive/Magang Prima PTKI/models/otomotif/scaler_X_automotive.pkl')
scaler_y = joblib.load('/content/drive/MyDrive/Magang Prima PTKI/models/otomotif/scaler_y_automotive.pkl')
```

```python
def get_forecast_7_days_stabilized_automotive(model, df_historical, scaler_X, scaler_y):
    # Ambil data 7 hari terakhir
    current_lags = df_historical['revenue_clean'].tail(7).tolist()

    # Hitung rata-rata per hari dalam seminggu (untuk Rem/Stabilizer)
    weekly_pattern = df_historical.groupby(df_historical.index.dayofweek)['revenue_clean'].mean()

    predictions = []
    dates = []
    last_date = df_historical.index[-1]
    id_holidays = holidays.Indonesia(years=[last_date.year, last_date.year + 1])

    print("="*60)
    print(f"   PREDIKSI OPERASIONAL AUTOMOTIVE TERSTABILISASI (7 HARI)")
    print(f"   (Batas Data: {last_date.strftime('%d %b %Y')})")
    print("="*60)

    for i in range(1, 8):
        current_date = last_date + timedelta(days=i)
        day_idx = current_date.weekday()

        # Fitur Engineering
        is_holiday = 1 if current_date.strftime('%Y-%m-%d') in id_holidays else 0
        is_payday = 1 if current_date.day >= 25 or current_date.day <= 5 else 0
        is_weekend = 1 if day_idx >= 5 else 0
        rolling_avg = np.mean(current_lags)
        day_of_month = current_date.day

        # Susun fitur (12 fitur)
        input_row = current_lags + [is_holiday, is_payday, is_weekend, rolling_avg, day_of_month]

        # Scaling & Prediksi
        input_scaled = scaler_X.transform([input_row])
        input_final = input_scaled.reshape((1, 1, len(features)))
        pred_scaled = model.predict(input_final, verbose=0)
        raw_pred = scaler_y.inverse_transform(pred_scaled)[0][0]

        # --- LOGIKA "REM" (STABILIZER) ---
        avg_hist_day = weekly_pattern[day_idx]
        stabilized_pred = (raw_pred * 0.7) + (avg_hist_day * 0.3)

        # Clipping: Mencegah angka tidak masuk akal
        max_limit = df_historical['revenue_clean'].max() * 1.2
        final_pred = min(stabilized_pred, max_limit)

        predictions.append(final_pred)
        dates.append(current_date)

        # Update Lags untuk iterasi berikutnya
        current_lags.pop(0)
        current_lags.append(final_pred)

        # Status berdasarkan statistik deskriptif data automotive
        status = "RAMAI" if final_pred > stats['mean'] * 1.1 else "NORMAL"
        print(f"{current_date.strftime('%A, %d %b')}: Rp {final_pred:,.0f} | {status}")

    return pd.DataFrame({'date': dates, 'pred_revenue': predictions}).set_index('date')
```

```python
# Jalankan Proyeksi Baru untuk Automotive
forecast_results_automotive = get_forecast_7_days_stabilized_automotive(best_model, time_series_automotive, scaler_X, scaler_y)
```

```python
# Visualisasi
plt.figure(figsize=(15, 6))
hist_data_auto = time_series_automotive.tail(30)

plt.plot(hist_data_auto.index, hist_data_auto['revenue_clean'],
         label='Data Aktual (Otomotif)', color='black', marker='o')

plt.plot(forecast_results_automotive.index, forecast_results_automotive['pred_revenue'],
         label='Prediksi AI Terstabilisasi', color='blue', linestyle='--', marker='s')

plt.axvline(x=hist_data_auto.index[-1], color='red', linestyle=':', label='Batas Hari Ini')
plt.title('Dashboard Forecasting Automotive (Dengan Logika Penyeimbang Mingguan)', fontsize=14)
plt.ylabel('Pendapatan (IDR)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

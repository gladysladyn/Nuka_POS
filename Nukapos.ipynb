### 1. provinces, cities, districts, villages

```python
# --- URLs API Wilayah ---
PROVINCES_URL = "https://wilayah.id/api/provinces.json"
REGENCIES_URL = "https://wilayah.id/api/regencies/{PROVINCE_CODE}.json"
DISTRICTS_URL = "https://wilayah.id/api/districts/{REGENCY_CODE}.json"
VILLAGES_URL = "https://wilayah.id/api/villages/{DISTRICT_CODE}.json"

# --- FUNGSI BARU UNTUK MENGAMBIL DATA ---

def generate_locations_from_api():
    """
    Mengambil data wilayah (provinces, cities/regencies, districts, villages)
    dari API wilayah.id.
    """
    now = datetime.now()
    provinces_data = []
    cities_data = []
    districts_data = []
    villages_data = []

    print("Memulai pengambilan data dari API (Provinces -> Regencies -> Districts -> Villages)...")

    # 1. Ambil Data Provinsi
    try:
        response = requests.get(PROVINCES_URL)
        response.raise_for_status()
        provinces_api = response.json().get('data', [])
    except requests.exceptions.RequestException as e:
        print(f"Error mengambil data provinsi: {e}")
        return pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

    provinces_to_process = provinces_api
    print(f"Berhasil mengambil {len(provinces_to_process)} provinsi).")

    for prov in provinces_to_process:
        prov_code_api = prov['code'] # Kode asli untuk URL (e.g., '11')
        prov_code_int = int(prov_code_api.replace('.', '')) # ID integer untuk DB (e.g., 11)

        provinces_data.append({
            'id': prov_code_int, # Menggunakan ID integer
            'name': prov['name'],
            'createdAt': now,
            'updatedAt': now
        })

        # 2. Ambil Data Kota/Kabupaten (Regencies)
        try:
            # Gunakan kode API asli untuk URL
            regencies_url = REGENCIES_URL.format(PROVINCE_CODE=prov_code_api)
            response = requests.get(regencies_url)
            response.raise_for_status()
            regencies_api = response.json().get('data', [])
        except requests.exceptions.RequestException as e:
            print(f"Peringatan: Error mengambil data kota/kabupaten untuk {prov['name']}: {e}")
            continue

        for reg in regencies_api:
            reg_code_api = reg['code'] # Kode asli untuk URL (e.g., '11.05')
            reg_code_int = int(reg_code_api.replace('.', '')) # ID integer untuk DB (e.g., 1105)

            cities_data.append({
                'id': reg_code_int, # Menggunakan ID integer
                'provinceId': prov_code_int, # Menggunakan ID integer
                'name': reg['name'],
                'createdAt': now,
                'updatedAt': now
            })

            # 3. Ambil Data Kecamatan (Districts)
            try:
                # Gunakan kode API asli untuk URL
                districts_url = DISTRICTS_URL.format(REGENCY_CODE=reg_code_api)
                response = requests.get(districts_url)
                response.raise_for_status()
                districts_api = response.json().get('data', [])
            except requests.exceptions.RequestException as e:
                print(f"Peringatan: Error mengambil data kecamatan untuk {reg['name']}: {e}")
                continue

            for dist in districts_api:
                dist_code_api = dist['code'] # Kode asli untuk URL (e.g., '11.05.01')
                dist_code_int = int(dist_code_api.replace('.', '')) # ID integer untuk DB (e.g., 110501)

                districts_data.append({
                    'id': dist_code_int, # Menggunakan ID integer
                    'cityId': reg_code_int, # Menggunakan ID integer
                    'name': dist['name'],
                    'createdAt': now,
                    'updatedAt': now
                })

                # 4. Ambil Data Desa/Kelurahan (Villages)
                try:
                    # Gunakan kode API asli untuk URL
                    villages_url = VILLAGES_URL.format(DISTRICT_CODE=dist_code_api)
                    response = requests.get(villages_url)
                    response.raise_for_status()
                    villages_api = response.json().get('data', [])
                except requests.exceptions.RequestException as e:
                    print(f"Peringatan: Error mengambil data desa/kelurahan untuk {dist['name']}: {e}")

                    # Jika gagal, masukkan 3 dummy villages
                    for i in range(3):
                        villages_data.append({
                            # Gunakan dist_code_int yang sudah bersih untuk ID fallback
                            'id': int(f'{dist_code_int}00{i+1}'),
                            'districtId': dist_code_int, # Menggunakan ID integer
                            'name': fake.city_prefix() + ' ' + fake.city_suffix(),
                            'createdAt': now,
                            'updatedAt': now
                        })
                    continue

                for village in villages_api:
                    village_code_api = village['code'] # Kode asli (e.g., '11.05.01.2001')
                    village_code_int = int(village_code_api.replace('.', '')) # ID integer (e.g., 1105012001)

                    villages_data.append({
                        'id': village_code_int, # Menggunakan ID integer
                        'districtId': dist_code_int, # Menggunakan ID integer
                        'name': village['name'],
                        'createdAt': now,
                        'updatedAt': now
                    })

    # Konversi ke DataFrame: Periksa jika data kosong sebelum konversi
    if provinces_data:
        # Jika data ada, buat DataFrame
        provinces_df = pd.DataFrame(provinces_data).astype({'id': 'int64'})
        cities_df = pd.DataFrame(cities_data).astype({'id': 'int64', 'provinceId': 'int64'})
        districts_df = pd.DataFrame(districts_data).astype({'id': 'int64', 'cityId': 'int64'})
        villages_df = pd.DataFrame(villages_data).astype({'id': 'int64', 'districtId': 'int64'})
    else:
        # Jika provinces_data kosong, buat DataFrame kosong dengan kolom yang benar
        provinces_df = pd.DataFrame(columns=['id', 'name', 'createdAt', 'updatedAt'])
        cities_df = pd.DataFrame(columns=['id', 'provinceId', 'name', 'createdAt', 'updatedAt'])
        districts_df = pd.DataFrame(columns=['id', 'cityId', 'name', 'createdAt', 'updatedAt'])
        villages_df = pd.DataFrame(columns=['id', 'districtId', 'name', 'createdAt', 'updatedAt'])

    print(f"\nSelesai. Total data yang diambil:")
    print(f"- Provinces: {len(provinces_df)}")
    print(f"- Cities: {len(cities_df)}")
    print(f"- Districts: {len(districts_df)}")
    print(f"- Villages: {len(villages_df)}")

    return provinces_df, cities_df, districts_df, villages_df

# --- EKSEKUSI ---
provinces_df, cities_df, districts_df, villages_df = generate_locations_from_api()

# --- SIMPAN KE CSV ---
provinces_df.to_csv('provinces.csv', index=False)
cities_df.to_csv('cities.csv', index=False)
districts_df.to_csv('districts.csv', index=False)
villages_df.to_csv('villages.csv', index=False)
print("\nData lokasi (provinces, cities, districts, villages) berhasil diperbarui secara lengkap dari API dan disimpan ke CSV.")
```

```python
# --- FUNGSI NAMA SPESIFIK GENDER ---
def generate_gender_name():
    """Menghasilkan nama depan dan belakang yang sesuai gender (L/P) secara acak."""
    if random.random() < 0.5:
        # 50% Laki-laki
        name = f"{fake.first_name_male()} {fake.last_name_male()}"
    else:
        # 50% Perempuan
        name = f"{fake.first_name_female()} {fake.last_name_female()}"
    return name

# --- FUNGSI UNTUK EMAIL ---
def generate_email(name, role_name, existing_emails):
    """Menghasilkan email unik berdasarkan nama pengguna."""

    # Bersihkan nama: ganti spasi dengan titik atau garis bawah
    clean_name = name.lower().replace(' ', random.choice(['.', '_', '']))

    # Daftar domain untuk personal/UMKM
    realistic_domains = ['gmail.com', 'yahoo.com', 'outlook.com', 'mail.com', 'bisnisanda.com', 'posku.id']

    # 30% menggunakan domain bisnis berdasarkan role (untuk Superadmin, Owner, Admin Toko)
    if role_name in ['Superadmin', 'Owner', 'Admin Toko'] and random.random() < 0.3:
        domain = f"nukapos-{role_name.lower().replace(' ', '')}.com"
    else:
        domain = random.choice(realistic_domains)

    # Menambahkan angka acak kecil untuk menjamin keunikan
    while True:
        suffix = random.choice(['', str(random.randint(1, 99))])
        email = f"{clean_name}{suffix}@{domain}"

        if email not in existing_emails:
            existing_emails.add(email)
            return email

# --- FUNGSI UNTUK ENKRIPSI PASSWORD ---
def hash_password(password):
    """Mengenkripsi password menggunakan bcrypt."""
    # Menggunakan salt default (otomatis dihasilkan)
    hashed = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())
    return hashed.decode('utf-8')

# --- Fungsi Previx Nomor Telepon ---
OPERATOR_PREFIXES = {
    'Telkomsel': ['0811', '0812', '0813', '0821', '0822', '0823', '0851', '0852', '0853'],
    'XL Axiata': ['0817', '0818', '0819', '0859', '0877', '0878'],
    'AXIS': ['0831', '0832', '0838'],
    'Indosat': ['0814', '0815', '0857', '0858', '0874'],
    'Tri (3)': [f'089{i}' for i in range(5, 10)],
    'Smartfren': ['0881', '0882', '0886', '0889']
}

# --- Fungsi Nomor Telepon ---
def generate_phone(existing_numbers):
    """Menghasilkan nomor telepon diawali dengan kode negara '62'."""
    while True:
        # Pilih operator dan prefix
        operator = random.choice(list(OPERATOR_PREFIXES.keys()))
        prefix_domestik = random.choice(OPERATOR_PREFIXES[operator])

        # Sisa panjang nomor antara 6 hingga 9 digit (Total 10-13 digit)
        remaining_length = random.randint(6, 9)
        remaining_numbers = fake.numerify('#' * remaining_length)

        full_number = '62' + prefix_domestik[1:] + remaining_numbers

        # Periksa keunikan
        if full_number not in existing_numbers:
            existing_numbers.add(full_number)
            return full_number


# --- FUNGSI UNTUK USERS ---
def generate_users(roles_df, num_owners=5, num_admin_toko=10, num_kasir=10):
    """
    Membuat data dummy untuk tabel users.
    """
    now = datetime.now()
    users_data = []
    generated_phone_numbers = set()
    generated_emails = set()

    # Ambil roleId dan hash_password
    superadmin_id = roles_df[roles_df['roleName'] == 'Superadmin']['id'].iloc[0]
    owner_id = roles_df[roles_df['roleName'] == 'Owner']['id'].iloc[0]
    admin_toko_id = roles_df[roles_df['roleName'] == 'Admin Toko']['id'].iloc[0]
    kasir_id = roles_df[roles_df['roleName'] == 'Kasir']['id'].iloc[0]
    default_password_hash = hash_password('password123')

    # --- 1. Superadmin ---
    name_sa = generate_gender_name()
    sa_phone_number = generate_phone(generated_phone_numbers)

    # Email Superadmin tetap statis
    sa_email = 'superadmin@nukapos.com'
    generated_emails.add(sa_email)

    users_data.append({
        'id': 1,
        'name': name_sa,
        'email': sa_email,
        'password': default_password_hash,
        'imageUrl': None,
        'phoneNumber': sa_phone_number,
        'isEmailVerified': 1,
        'idCard': fake.unique.numerify('################'),
        'idCardImage': 'ktp_superadmin.jpg',
        'npwp': fake.unique.numerify('################'),
        'npwpImage': 'npwp_superadmin.jpg',
        'googleId': None,
        'roleId': superadmin_id,
        'createdAt': now,
        'updatedAt': now
    })

    user_id_counter = 2

    # --- 2. Owners ---
    for i in range(num_owners):
        name = generate_gender_name()
        users_data.append({
            'id': user_id_counter,
            'name': name,
            'email': generate_email(name, 'Owner', generated_emails),
            'password': default_password_hash,
            'imageUrl': fake.image_url(),
            'phoneNumber': generate_phone(generated_phone_numbers),
            'isEmailVerified': 1,
            'idCard': fake.unique.numerify('################'),
            'idCardImage': f'ktp_owner_{i+1}.jpg',
            'npwp': fake.unique.numerify('################'),
            'npwpImage': f'npwp_owner_{i+1}.jpg',
            'googleId': None,
            'roleId': owner_id,
            'createdAt': now,
            'updatedAt': now
        })
        user_id_counter += 1

    # --- 3. Karyawan: Admin Toko ---
    for i in range(num_admin_toko):
        name = generate_gender_name()
        users_data.append({
            'id': user_id_counter,
            'name': name,
            'email': generate_email(name, 'Admin Toko', generated_emails),
            'password': default_password_hash,
            'imageUrl': fake.image_url(),
            'phoneNumber': generate_phone(generated_phone_numbers),
            'isEmailVerified': random.choice([0, 1]),
            'idCard': fake.unique.numerify('################'),
            'idCardImage': f'ktp_admintoko_{i+1}.jpg',
            'npwp': None if random.random() < 0.7 else fake.unique.numerify('################'),
            'npwpImage': None,
            'googleId': None,
            'roleId': admin_toko_id,
            'createdAt': now,
            'updatedAt': now
        })
        user_id_counter += 1

    # --- 4. Karyawan: Kasir ---
    for i in range(num_kasir):
        name = generate_gender_name()
        users_data.append({
            'id': user_id_counter,
            'name': name,
            'email': generate_email(name, 'Kasir', generated_emails),
            'password': default_password_hash,
            'imageUrl': fake.image_url(),
            'phoneNumber': generate_phone(generated_phone_numbers),
            'isEmailVerified': random.choice([0, 1]),
            'idCard': fake.unique.numerify('################'),
            'idCardImage': f'ktp_kasir_{i+1}.jpg',
            'npwp': None if random.random() < 0.7 else fake.unique.numerify('################'),
            'npwpImage': None,
            'googleId': None,
            'roleId': kasir_id,
            'createdAt': now,
            'updatedAt': now
        })
        user_id_counter += 1

    users_df = pd.DataFrame(users_data)
    users_df['id'] = users_df['id'].astype(np.int64)
    users_df['roleId'] = users_df['roleId'].astype(np.int64)
    return users_df

# --- EKSEKUSI ---
users_df = generate_users(roles_df, num_owners=5, num_admin_toko=10, num_kasir=10)
users_df.to_csv('users.csv', index=False)
print("Data users.csv berhasil dibuat.")
```

```python
# --- FUNGSI GENERATE TENANTS ---
def generate_tenants(owner_users_df, store_types_df, balikpapan_villages_df, balikpapan_districts_df, kaltim_id, balikpapan_city_id):
    """
    Membuat data dummy untuk tabel tenants dengan:
    1. Nama Tenant = Nama Pemilik (UMKM Perorangan).
    2. NPWP diambil dari data user (pribadi) jika ada.
    """
    now = datetime.now()
    tenants_data = []

    # Gabungkan Owner Users dan Tenants Data
    tenant_creation_data = owner_users_df.copy() # Mulai dari data Owner

    for i in range(len(tenant_creation_data)):
        owner_data = tenant_creation_data.iloc[i]

        # 1. Ambil Lokasi Geografis Acak DARI BALIKPAPAN
        random_village = balikpapan_villages_df.sample(1).iloc[0]
        village_id = random_village['id']
        district_id = random_village['districtId']

        # Province ID dan City ID di-hardcode ke Balikpapan ID
        province_id = kaltim_id
        city_id = balikpapan_city_id

        # 2. Ambil Owner ID
        owner_user_id = owner_data['id']

        # 3. Nama Tenant: Menggunakan Nama Pemilik (user.name)
        tenant_name = owner_data['name']

        # 4. NPWP: Menggunakan NPWP Pemilik (user.npwp)
        npwp_number = owner_data['npwp']
        npwp_image = owner_data['npwpImage'].replace('owner', 'tenant') if pd.notna(owner_data['npwpImage']) else None

        tenants_data.append({
            'id': i + 1,
            'userId': owner_user_id, # Owner (UNIQUE Foreign Key)
            'name': tenant_name,
            'npwp': npwp_number,
            'npwpImage': npwp_image,
            'address': f"{fake.street_address()}, Kel. {random_village['name']}",
            'provinceId': int(province_id),
            'cityId': int(city_id),
            'districtId': int(district_id),
            'villageId': int(village_id),
            'postalCode': fake.numerify('#####'), # 5 Digit Angka Acak
            'tenantStatus': 'ACTIVE', # ENUM: ACTIVE, INACTIVE
            'createdAt': now - timedelta(days=random.randint(10, 30)),
            'updatedAt': now
        })

    tenants_df = pd.DataFrame(tenants_data)
    # Sesuaikan tipe data integer
    tenants_df['id'] = tenants_df['id'].astype(np.int64)
    tenants_df['provinceId'] = tenants_df['provinceId'].astype(np.int64)
    tenants_df['cityId'] = tenants_df['cityId'].astype(np.int64)
    tenants_df['districtId'] = tenants_df['districtId'].astype(np.int64)
    tenants_df['villageId'] = tenants_df['villageId'].astype(np.int64)

    return tenants_df

# --- EKSEKUSI ---
tenants_df = generate_tenants(owner_users_df, store_types_df, balikpapan_villages_df, balikpapan_districts_df, kaltim_id, balikpapan_city_id)
tenants_df.to_csv('tenants.csv', index=False)
print(f"Data tenants.csv berhasil dibuat. Total {len(tenants_df)} tenants.")
```

### 2. tokens

```python
# --- FUNGSI GENERATE TOKENS ---
def generate_tokens(users_df):
    """
    Membuat data dummy untuk tabel tokens (access dan refresh token)
    untuk semua user.
    """
    now = datetime.now()
    tokens_data = []
    token_id_counter = 1

    # Tipe Token: ACCESS, REFRESH, RESET_PASSWORD, VERIFY_EMAIL
    token_types = ['ACCESS', 'REFRESH', 'VERIFY_EMAIL']

    for user_id in users_df['id'].tolist():
        for token_type in token_types:

            # Waktu kadaluarsa berbeda untuk setiap tipe
            if token_type == 'ACCESS':
                expires_at = now + timedelta(hours=random.randint(1, 6))
                is_blacklisted = 0 # Access token jarang di-blacklist
            elif token_type == 'REFRESH':
                expires_at = now + timedelta(days=random.randint(7, 30))
                is_blacklisted = random.choice([0, 0, 0, 1]) # Kadang di-blacklist
            else: # VERIFY_EMAIL
                expires_at = now + timedelta(days=7)
                is_blacklisted = 0

            tokens_data.append({
                'id': token_id_counter,
                'userId': user_id,
                'token': fake.uuid4(), # Token string unik
                'type': token_type, # ENUM
                'expires': expires_at,
                'blacklisted': is_blacklisted, # TINYINT(1)
                'createdAt': now - timedelta(hours=random.randint(1, 50)),
            })
            token_id_counter += 1

    tokens_df = pd.DataFrame(tokens_data)
    return tokens_df

# --- EKSEKUSI ---
tokens_df = generate_tokens(users_df)
tokens_df.to_csv('tokens.csv', index=False)
print(f"Data tokens.csv berhasil dibuat. Total {len(tokens_df)} tokens.")
```

### 3. tenant_users

```python
# --- FUNGSI GENERATE TENANT USERS ---
def generate_tenant_users(tenants_df, users_df, roles_df):
    """
    Membuat relasi antara tenants dan users (Owner dan Superadmin).
    """
    now = datetime.now()
    tenant_users_data = []

    # 1. Cari Role ID Superadmin dari roles_df
    superadmin_role_id = roles_df[roles_df['roleName'] == 'Superadmin']['id'].iloc[0]
    # 2. Cari User ID Superadmin dari users_df menggunakan roleId
    superadmin_id = users_df[users_df['roleId'] == superadmin_role_id]['id'].iloc[0]

    user_tenant_counter = 1

    for _, tenant in tenants_df.iterrows():
        tenant_id = tenant['id']
        owner_id = tenant['userId'] # Owner

        # 1. Hubungkan Owner dengan Tenant-nya
        tenant_users_data.append({
            'id': user_tenant_counter,
            'userId': owner_id,
            'tenantId': tenant_id,
            'status': 'ACTIVE', # ENUM: ACTIVE, INACTIVE
            'createdAt': now - timedelta(days=random.randint(5, 25)),
            'updatedAt': now
        })
        user_tenant_counter += 1

        # 2. Hubungkan Superadmin dengan SEMUA Tenant
        tenant_users_data.append({
            'id': user_tenant_counter,
            'userId': superadmin_id,
            'tenantId': tenant_id,
            'status': 'ACTIVE',
            'createdAt': now - timedelta(days=random.randint(5, 25)),
            'updatedAt': now
        })
        user_tenant_counter += 1

    tenant_users_df = pd.DataFrame(tenant_users_data)
    # Sesuaikan tipe data
    tenant_users_df['id'] = tenant_users_df['id'].astype(np.int64)
    tenant_users_df['userId'] = tenant_users_df['userId'].astype(np.int64)
    tenant_users_df['tenantId'] = tenant_users_df['tenantId'].astype(np.int64)
    return tenant_users_df

# --- EKSEKUSI ---
tenant_users_df = generate_tenant_users(tenants_df, users_df, roles_df)
tenant_users_df.to_csv('tenant_users.csv', index=False)
print(f"Data tenant_users.csv berhasil dibuat. Total {len(tenant_users_df)} relasi.")
```

```python
# --- FUNGSI GENERATE SUPPLIERS (Tahap 3.4) ---
def generate_suppliers(tenants_df, users_df, generated_phone_numbers):
    """
    Membuat data dummy untuk tabel suppliers.
    Mayoritas supplier diikat ke tenant tertentu (70%), sisanya global (30%).
    """
    now = datetime.now()
    suppliers_data = []

    tenant_ids = tenants_df['id'].tolist()
    NUM_SUPPLIERS = 20

    # Prefix legalitas realistis
    prefix_supplier = ['CV.', 'PT.', 'UD.', 'Koperasi', 'Distributor']
    # Kata kunci umum pengganti nama belakang
    brand_words = ['Mitra', 'Nusantara', 'Prima', 'Cipta', 'Berkah', 'Sinar', 'Global', 'Lestari', 'Bintang', 'Sakti']
    # Suffix deskriptif/realistis
    suffix_supplier = ['Logistik', 'Perkasa', 'Sejati', 'Mandiri', 'Unggul', 'Jaya', 'Pangan']

    for i in range(1, NUM_SUPPLIERS + 1):
        # 70% diikat ke tenant (lokal), 30% NULL (global)
        tenant_id = random.choice(tenant_ids) if random.random() < 0.7 else None

        # Penamaan Supplier yang realistis (Prefix + Kata Brand + Suffix)
        supplier_name = f"{random.choice(prefix_supplier)} {random.choice(brand_words)} {random.choice(suffix_supplier)}"

        suppliers_data.append({
            'id': i,
            'name': supplier_name,
            'tenantId': tenant_id, # NULLABLE
            'phoneNumber': generate_phone(generated_phone_numbers),
            'email': generate_email(supplier_name.replace('.', '').replace(' ', ''), 'Supplier', set()),
            'address': fake.address(),
            'createdAt': now - timedelta(days=random.randint(1, 100)), # Variasi Waktu
            'updatedAt': now
        })

    suppliers_df = pd.DataFrame(suppliers_data)
    # Sesuaikan tipe data ID (INT(11))
    suppliers_df['id'] = suppliers_df['id'].astype(np.int64)
    # Menggunakan 'Int64' Pandas untuk menampung nilai NaN/None
    suppliers_df['tenantId'] = suppliers_df['tenantId'].astype('Int64')

    return suppliers_df

# --- EKSEKUSI ---
generated_phone_numbers = set(users_df['phoneNumber'].tolist())

suppliers_df = generate_suppliers(tenants_df, users_df, generated_phone_numbers)
suppliers_df.to_csv('suppliers.csv', index=False)
print(f"Data suppliers.csv berhasil dibuat. Total {len(suppliers_df)} suppliers.")
```

### 5. stores

```python
# --- FUNGSI GENERATE STORES ---
# Total: 1 + 3 + 2 + 2 + 2 = 10 Toko
STORE_COUNT_DISTRIBUTION = [1, 3, 2, 2, 2]

# Data Suffix (Tetap)
suffix_umum = ['Jaya', 'Barokah', 'Makmur', 'Abadi', 'Sejahtera', 'Bersama', 'Mandiri', 'Unggul', 'Sakti']

# --- MAPPING PREFIX BERDASARKAN TIPE TOKO ---
PREFIX_MAPPING_BY_TYPE = {
    # Makanan & Minuman
    1: ['Warung', 'Kedai', 'Kantin', 'Dapur', 'Restoran'],
    # Fashion & Aksesoris
    2: ['Toko', 'Butik', 'Gallery', 'Outlet'],
    # Perawatan Diri & Kecantikan
    3: ['Toko', 'Salon', 'Griya', 'Pusat'],
    # Kebutuhan Kesehatan
    4: ['Apotek', 'Klinik', 'Toko Obat', 'Pusat'],
    # Rumah Tangga & Dekorasi
    5: ['Toko', 'Grosir', 'Pusat', 'Bazaar'],
    # Peralatan Bayi & Anak
    6: ['Toko', 'Grosir', 'Pusat'],
    # Mainan, Hobi & Peralatan Outdoor
    7: ['Toko', 'Kios', 'Pusat'],
    # Elektronik & Gadget
    8: ['Toko', 'Outlet', 'Pusat'],
    # Pertanian, Perkebunan & Peternakan
    9: ['Toko', 'Grosir', 'Gudang', 'Kios'],
    # Otomotif & Suku Cadang
    10: ['Bengkel', 'Dealer', 'Toko', 'Pusat Suku Cadang'],
    # Perlengkapan Hewan Peliharaan
    11: ['Toko', 'Pet Shop', 'Grosir'],
    # Minimarket & Kelontong
    12: ['Minimarket', 'Kios', 'Toko', 'Warung'],
}

def get_type_specific_prefix(store_type_id):
    """Mengambil prefix acak yang cocok berdasarkan store type ID."""
    # Mengambil list prefix berdasarkan ID, atau fallback ke list umum jika ID tidak terdaftar
    prefix_list = PREFIX_MAPPING_BY_TYPE.get(store_type_id, ['Toko', 'Pusat'])
    return random.choice(prefix_list)

# --- FUNGSI GENERATE STORES ---
def generate_stores(tenants_df, store_types_df, balikpapan_villages_df, balikpapan_districts_df, kaltim_id, balikpapan_city_id, generated_phone_numbers):
    """
    Membuat data dummy untuk tabel stores.
    Nama toko dibuat SINKRON dengan Store Type ID yang ditetapkan (PREFIX TERIKAT).
    """
    now = datetime.now()
    stores_data = []
    store_id_counter = 1
    generated_store_codes = set()

    # Map untuk lookup cepat Store Type Name
    store_type_map = store_types_df.set_index('id')['name'].to_dict()
    store_type_ids = store_types_df['id'].tolist()

    def get_simplified_theme(store_type_id):
        return store_type_map[store_type_id].split('&')[0].strip()

    if len(STORE_COUNT_DISTRIBUTION) != len(tenants_df):
        store_counts = [random.randint(1, 3) for _ in range(len(tenants_df))]
    else:
        store_counts = STORE_COUNT_DISTRIBUTION


    for idx, tenant in tenants_df.iterrows():
        tenant_id = tenant['id']

        num_stores_for_tenant = store_counts[idx]

        # --- 1. PRE-SELEKSI STORE TYPE ID (Decide ID first) ---
        is_homogeneous = random.random() < 0.7 and num_stores_for_tenant > 1

        store_type_list = []
        base_brand_name = None

        if is_homogeneous or num_stores_for_tenant == 1:
            # Pilih satu tipe dasar untuk semua toko tenant ini
            base_store_type_id = random.choice(store_type_ids)
            store_type_list = [base_store_type_id] * num_stores_for_tenant

            # Buat Nama Brand Dasar (SINKRON dengan ID dan PREFIX TERIKAT)
            base_prefix = get_type_specific_prefix(base_store_type_id)
            store_type_theme = get_simplified_theme(base_store_type_id)
            base_brand_name = f"{base_prefix} {store_type_theme} {random.choice(suffix_umum)}"

        else:
            # Diversifikasi: ID berbeda-beda
            if len(store_type_ids) >= num_stores_for_tenant:
                store_type_list = random.sample(store_type_ids, num_stores_for_tenant)
            else:
                store_type_list = [random.choice(store_type_ids) for _ in range(num_stores_for_tenant)]

        # Ambil lokasi acak di Balikpapan
        random_village = balikpapan_villages_df.sample(1).iloc[0]
        village_id = random_village['id']
        district_id = random_village['districtId']

        province_id = kaltim_id
        city_id = balikpapan_city_id

        store_emails = set()

        for j in range(num_stores_for_tenant):
            store_type_id = store_type_list[j]
            current_store_id = store_id_counter
            store_code = f"NUKA-BP-{current_store_id:03d}"
            generated_store_codes.add(store_code)

            # 2. Nama Toko: Logika Bersyarat
            if num_stores_for_tenant == 1:
                store_name = base_brand_name # Base brand sudah sinkron
            elif is_homogeneous:
                store_name = f"{base_brand_name} Cabang {chr(65 + j)}" # Base brand sudah sinkron
            else:
                # Diversifikasi: Generate Brand Name Unik (SINKRON dengan store_type_id)
                current_prefix = get_type_specific_prefix(store_type_id) # PREFIX TERIKAT
                current_theme = get_simplified_theme(store_type_id)
                store_name = f"{current_prefix} {current_theme} {random.choice(suffix_umum)}"

            # 3. Email Toko
            store_email = generate_email(store_name.replace(' ', '').replace('.', ''), 'Store', store_emails)

            stores_data.append({
                'id': current_store_id,
                'storeCode': store_code,
                'tenantId': tenant_id,
                'name': store_name,
                'address': f"{fake.street_address()}, Kel. {random_village['name']}, Kec. {balikpapan_districts_df[balikpapan_districts_df['id']==district_id]['name'].iloc[0]}",
                'email': store_email,
                'image': fake.image_url(),
                'provinceId': int(province_id),
                'cityId': int(city_id),
                'districtId': int(district_id),
                'villageId': int(village_id),
                'storeTypeId': store_type_id, # GUARANTEED SINKRON
                'postalCode': fake.numerify('#####'),
                'description': fake.paragraph(nb_sentences=3),
                'phoneNumber': generate_phone(generated_phone_numbers),
                'logo': fake.image_url(),
                'permitCertificate': f'izin_store_{current_store_id}.pdf' if random.random() < 0.5 else None,
                'status': 'ACTIVE',
                'createdAt': now - timedelta(days=random.randint(5, 20)),
                'updatedAt': now
            })
            store_id_counter = current_store_id + 1

    stores_df = pd.DataFrame(stores_data)
    # Sesuaikan tipe data integer (INT(11))
    stores_df['id'] = stores_df['id'].astype(np.int64)
    stores_df['tenantId'] = stores_df['tenantId'].astype(np.int64)
    stores_df['provinceId'] = stores_df['provinceId'].astype(np.int64)
    stores_df['cityId'] = stores_df['cityId'].astype(np.int64)
    stores_df['districtId'] = stores_df['districtId'].astype(np.int64)
    stores_df['villageId'] = stores_df['villageId'].astype(np.int64)
    stores_df['storeTypeId'] = stores_df['storeTypeId'].astype(np.int64)

    return stores_df

# --- EKSEKUSI ---
stores_df = generate_stores(tenants_df, store_types_df, balikpapan_villages_df, balikpapan_districts_df, kaltim_id, balikpapan_city_id, generated_phone_numbers)
stores_df.to_csv('stores.csv', index=False)
print(f"Data stores.csv berhasil dibuat. Total {len(stores_df)} toko di Balikpapan.")
```

### 6. store_users

```python
# --- FUNGSI GENERATE STORE USERS ---

def generate_store_users(stores_df, users_df, roles_df):
    """
    Membuat relasi many-to-many antara users (Admin Toko & Kasir) dan stores.
    """
    now = datetime.now()
    store_users_data = []
    store_user_id_counter = 1

    # Filter User Karyawan IDs
    admin_toko_id = roles_df[roles_df['roleName'] == 'Admin Toko']['id'].iloc[0]
    kasir_id = roles_df[roles_df['roleName'] == 'Kasir']['id'].iloc[0]

    admin_toko_users = users_df[users_df['roleId'] == admin_toko_id]['id'].tolist()
    kasir_users = users_df[users_df['roleId'] == kasir_id]['id'].tolist()

    # Store IDs (10 stores)
    stores_list = stores_df['id'].tolist()

    # Shuffle lists for random, but mandatory, assignment
    random.shuffle(admin_toko_users)
    random.shuffle(kasir_users)

    # Set untuk melacak relasi (userId, storeId) yang sudah ditugaskan
    assigned_relations = set()

    # --- 1. Fase Penugasan Wajib (1:1 Mapping) ---
    # Memastikan 10 toko mendapatkan 10 Admin Toko dan 10 Kasir yang unik

    for i, store_id in enumerate(stores_list):
        # Penugasan Admin Toko
        admin_user_id = admin_toko_users[i]

        # Penugasan Kasir
        kasir_user_id = kasir_users[i]

        # Tambahkan Admin Toko
        relation_key_admin = (admin_user_id, store_id)
        store_users_data.append({
            'id': store_user_id_counter,
            'userId': admin_user_id,
            'storeId': store_id,
            'status': 'ACTIVE',
            'createdAt': now - timedelta(days=random.randint(1, 10)),
            'updatedAt': now
        })
        assigned_relations.add(relation_key_admin)
        store_user_id_counter += 1

        # Tambahkan Kasir
        relation_key_kasir = (kasir_user_id, store_id)
        store_users_data.append({
            'id': store_user_id_counter,
            'userId': kasir_user_id,
            'storeId': store_id,
            'status': 'ACTIVE',
            'createdAt': now - timedelta(days=random.randint(1, 10)),
            'updatedAt': now
        })
        assigned_relations.add(relation_key_kasir)
        store_user_id_counter += 1

    # --- 2. Fase Penugasan Tambahan (Simulasi Fleksibilitas Staf) ---
    # Tambahkan 5 penugasan ekstra, di mana staf yang sama mungkin muncul di toko berbeda

    all_staff_users = admin_toko_users + kasir_users

    for _ in range(5):
        random_store_id = random.choice(stores_list)
        random_staff_id = random.choice(all_staff_users)
        relation_key = (random_staff_id, random_store_id)

        # Hanya tambahkan jika relasi user-store belum ada
        if relation_key not in assigned_relations:
             store_users_data.append({
                'id': store_user_id_counter,
                'userId': random_staff_id,
                'storeId': random_store_id,
                'status': random.choice(['ACTIVE', 'INACTIVE']), # Beberapa mungkin sudah non-aktif
                'createdAt': now - timedelta(days=random.randint(1, 15)),
                'updatedAt': now
            })
             assigned_relations.add(relation_key)
             store_user_id_counter += 1


    store_users_df = pd.DataFrame(store_users_data)
    # Sesuaikan tipe data
    store_users_df['id'] = store_users_df['id'].astype(np.int64) # BIGINT(20)
    store_users_df['userId'] = store_users_df['userId'].astype(np.int64) # BIGINT(20)
    store_users_df['storeId'] = store_users_df['storeId'].astype(np.int64) # INT(11)

    return store_users_df

    print("Memuat data roles, users, dan stores...")
try:
    roles_df = pd.read_csv('roles.csv')
    users_df = pd.read_csv('users.csv')
    stores_df = pd.read_csv('stores.csv')
except FileNotFoundError as e:
    print(f"Error: Salah satu file CSV (roles.csv, users.csv, atau stores.csv) tidak ditemukan: {e}")
    exit()


# --- EKSEKUSI ---
store_users_df = generate_store_users(stores_df, users_df, roles_df)
store_users_df.to_csv('store_users.csv', index=False)
print(f"Data store_users.csv berhasil dibuat. Total {len(store_users_df)} relasi staf-toko (minimal 20 relasi wajib, ditambah 5 acak).")
```

```python
# --- MAPPING GLOBAL OPTIONS ---
options_for_all_types = {
    'Ukuran': ['S', 'M', 'L', 'XL', 'XXL'],
    'Warna': ['Merah', 'Biru', 'Hijau', 'Hitam', 'Putih', 'Kuning', 'Abu-abu'],
    'Rasa': ['Original', 'Cokelat', 'Strawberry', 'Vanila', 'Keju', 'Pedas'],
    'Bahan': ['Katun', 'Poliester', 'Linen', 'Denim', 'Sintetis'],
    'Tipe Kemasan': ['Botol Plastik', 'Sachet', 'Dus Karton', 'Kaleng'],
    'Dosis': ['100mg', '250mg', '500mg'],
    'Jenis Obat': ['Tablet', 'Sirup', 'Kapsul'],
    'Material': ['Kayu', 'Besi', 'Plastik Keras', 'Kaca', 'Keramik'],
    'Merek': ['Brand A', 'Brand B', 'Brand C', 'Brand D'],
    'Tipe': ['Digital', 'Analog', 'Mekanik', 'Semi-Otomatis'],
    'Kapasitas': ['4GB', '8GB', '16GB', '32GB', '64GB'],
    'Berat': ['100gr', '250gr', '500gr', '1kg', '5kg'],
    'Jenis Tanaman': ['Sayur', 'Buah', 'Hias'],
    'Jenis': ['Matic', 'Manual', 'Elektrik', 'Jasa Servis'],
    'Aroma': ['Floral', 'Fruity', 'Woody', 'Fresh'],
    'Kandungan': ['Vitamin C', 'Kolagen', 'Hyaluronic Acid']
}

# --- MAPPING CUSTOM OPTIONS UNTUK VARIANT STORE ---
CUSTOM_OPTIONS_MAP = {
    'Level Pedas': ['Level 1 (Mild)', 'Level 3 (Hot)', 'Level 5 (Extrem)'],
    'Tambahan': ['Ya', 'Tidak'], # Untuk 'Topping Tambahan', 'Bordir Tambahan', dsb.
    'Spesial': ['Set B', 'Set C', 'Set D'], # Untuk 'Promo Beli 2', 'Set Khusus', dsb.
    'Kustom': ['Opsi A', 'Opsi B', 'Opsi C'], # Untuk Varian kustom umum
}

# --- FUNGSI GENERATE VARIANT OPTIONS (Tahap 3.8) ---
def generate_variant_options(variant_types_df, options_map, custom_options_map):
    """
    Membuat data master untuk tabel variant_options.
    Harga tambahan hanya ditetapkan pada opsi PREMIUM spesifik.
    """
    now = datetime.now()
    variant_options_data = []
    option_id_counter = 1

    for _, variant_type in variant_types_df.iterrows():
        type_id = variant_type['id']
        type_name = variant_type['name']
        type_scope = variant_type['scope']

        options_list = []

        # [Logika penetapan options_list (Tidak Berubah)]
        if type_scope == 'GLOBAL':
            options_list = options_map.get(type_name, [])
        elif type_scope == 'STORE':
            base_name = type_name.split('(')[0].strip() if '(' in type_name else type_name
            if 'Level Pedas' in base_name:
                options_list = custom_options_map['Level Pedas']
            elif 'Tambahan' in base_name or 'Bordir' in base_name or 'Khusus' in base_name:
                options_list = custom_options_map['Tambahan']
            elif 'Promo' in base_name or 'Edisi' in base_name or 'Set' in base_name:
                 options_list = custom_options_map['Spesial']
            else:
                 options_list = custom_options_map['Kustom']


        for option_name in options_list:

            # --- LOGIKA HARGA AKHIR ---

            # 1. Identifikasi Opsi Premium
            is_premium_size_weight = option_name.endswith('XXL') or option_name.endswith('5kg')
            is_premium_service = option_name == 'Jasa Servis'

            if is_premium_size_weight:
                # Harga premium untuk ukuran/berat (misal: Rp15,000 - Rp45,000)
                additional_price = random.randint(15000, 45000)
                additional_price = round(additional_price, -3)
            elif is_premium_service:
                # Harga premium untuk jasa (misal: Rp50,000 - Rp150,000)
                additional_price = random.randint(50000, 150000)
                additional_price = round(additional_price, -3)
            else:
                # 2. Default ke Nol (Harga tambahan kecil dihilangkan)
                additional_price = 0
            # -------------------------

            variant_options_data.append({
                'id': option_id_counter,
                'variantTypeId': type_id,
                'name': option_name,
                'additionalPrice': additional_price,
                'status': random.choice(['ACTIVE', 'INACTIVE']),
                'createdAt': now - timedelta(hours=random.randint(1, 48)),
                'updatedAt': now,
            })
            option_id_counter += 1

    variant_options_df = pd.DataFrame(variant_options_data)

    # --- SESUAIKAN TIPE DATA BIGINT(20) ---
    variant_options_df['id'] = variant_options_df['id'].astype(np.int64)
    variant_options_df['variantTypeId'] = variant_options_df['variantTypeId'].astype(np.int64)
    variant_options_df['additionalPrice'] = variant_options_df['additionalPrice'].astype(np.int64)

    return variant_options_df

# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data variant_types...")
try:
    variant_types_df = pd.read_csv('variant_types.csv')
except FileNotFoundError as e:
    print(f"Error: variant_types.csv tidak ditemukan. Pastikan Tahap 3.7 sudah dieksekusi: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE VARIANT OPTIONS ---
variant_options_df = generate_variant_options(variant_types_df, options_for_all_types, CUSTOM_OPTIONS_MAP)

# --- SIMPAN DAN TAMPILKAN HASIL ---
variant_options_df.to_csv('variant_options.csv', index=False)
print(f"Data variant_options.csv berhasil dibuat. Total {len(variant_options_df)} opsi varian.")
```

### 2. supplier_products

```python
# --- FUNGSI GENERATE SUPPLIER PRODUCTS (Tahap 4.2) ---
def generate_supplier_products(products_df, suppliers_df):
    """
    Membuat relasi many-to-many antara products dan suppliers.
    Logika: Produk harga rendah (<= 40k) memiliki 3-5 supplier, yang lain 1-2 supplier.
    """
    now = datetime.now()
    supplier_products_data = []
    rel_id_counter = 1
    supplier_ids = suppliers_df['id'].tolist()

    # Menentukan role yang mengelola produk (ENUM: ADMIN_TOKO, ADMIN_SCM, SUPER_SCM)
    managed_by_roles = ['ADMIN_TOKO', 'ADMIN_SCM', 'SUPER_SCM']

    for _, product in products_df.iterrows():
        product_id = product['id']
        base_price = product['basePrice']

        # Logika Kuantitas Supplier:
        if base_price <= 40000:
            # Produk harga rendah/umum: 3 hingga 5 supplier
            num_suppliers = random.randint(3, 5)
        else:
            # Produk harga standar/khusus: 1 hingga 2 supplier
            num_suppliers = random.randint(1, 2)

        # Pilih supplier secara acak tanpa pengulangan
        selected_suppliers = random.sample(supplier_ids, k=min(num_suppliers, len(supplier_ids)))

        for supplier_id in selected_suppliers:
            # Buat relasi untuk tabel supplier_products
            supplier_products_data.append({
                'id': rel_id_counter,
                'supplierId': supplier_id, # INT(11)
                'productId': product_id, # BIGINT(20)
                'managedBy': random.choice(managed_by_roles), # ENUM
            })
            rel_id_counter += 1

    supplier_products_df = pd.DataFrame(supplier_products_data)

    # Sesuaikan tipe data
    supplier_products_df['id'] = supplier_products_df['id'].astype(np.int64)
    supplier_products_df['supplierId'] = supplier_products_df['supplierId'].astype(np.int64)
    supplier_products_df['productId'] = supplier_products_df['productId'].astype(np.int64)

    return supplier_products_df

# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data products dan suppliers...")
try:
    products_df = pd.read_csv('products.csv')
    suppliers_df = pd.read_csv('suppliers.csv')
except FileNotFoundError as e:
    print(f"Error: Salah satu file CSV tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE SUPPLIER PRODUCTS ---
supplier_products_df = generate_supplier_products(products_df, suppliers_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
supplier_products_df.to_csv('supplier_products.csv', index=False)
print(f"Data supplier_products.csv berhasil dibuat. Total {len(supplier_products_df)} relasi produk-pemasok.")
```

```python
# --- FUNGSI GENERATE _PRODUCTTOSUPPLIER ---
def generate_producttosupplier(supplier_products_df):
    """
    Mentransformasi data dari supplier_products_df menjadi format junction table Prisma:
    A (productId) dan B (supplierId).
    """

    # Ambil kolom yang dibutuhkan dan ganti nama
    producttosupplier_df = supplier_products_df[[
        'productId',
        'supplierId'
    ]].rename(columns={
        'productId': 'A', # Referensi ke products.id (BIGINT(20))
        'supplierId': 'B'  # Referensi ke suppliers.id (INT(11))
    })

    # Sesuaikan tipe data sesuai skema
    producttosupplier_df['A'] = producttosupplier_df['A'].astype(np.int64)
    # Gunakan Int64 untuk memastikan kompatibilitas dengan INT(11) dan konsistensi Pandas
    producttosupplier_df['B'] = producttosupplier_df['B'].astype(np.int64)

    return producttosupplier_df


print("Memuat data supplier_products...")
try:
    # supplier_products_df harus dimuat dari Tahap 4.2
    supplier_products_df = pd.read_csv('supplier_products.csv')
except FileNotFoundError as e:
    print(f"Error: supplier_products.csv tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE _PRODUCTTOSUPPLIER ---
producttosupplier_df = generate_producttosupplier(supplier_products_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
producttosupplier_df.to_csv('_producttosupplier.csv', index=False)
print(f"Data _producttosupplier.csv berhasil dibuat. Total {len(producttosupplier_df)} relasi.")
```

### 4. product_variants

```python
# --- FUNGSI GENERATE PRODUCT VARIANTS ---

def generate_product_variants(products_df):
    """
    Membuat varian untuk sekitar 15% dari total produk.
    Setiap varian memiliki stok dan harga final (sementara disamakan dengan basePrice).
    """
    now = datetime.now()
    variants_data = []
    variant_id_counter = 1

    # Pilih 15% produk yang akan memiliki varian
    NUM_PRODUCTS_TO_VARIATE = int(len(products_df) * 0.15)
    products_with_variants = products_df.sample(n=NUM_PRODUCTS_TO_VARIATE, random_state=42)

    for _, product in products_with_variants.iterrows():
        product_id = product['id']
        base_price = product['basePrice']

        # Setiap produk memiliki 2 hingga 5 varian berbeda
        num_variants = random.randint(2, 5)

        for i in range(num_variants):

            # Stock: Antara 10 hingga 150
            stock = random.randint(10, 150)

            # Final Price: Untuk saat ini, asumsikan sama dengan Base Price (Harga akan disesuaikan di Tahap 4.5)
            final_price = base_price

            variants_data.append({
                'id': variant_id_counter,
                'productId': product_id, # BIGINT(20)
                'stock': stock, # INT(11)
                'finalPrice': final_price, # BIGINT(20)
                'isActive': 1,
                'createdAt': now - timedelta(days=random.randint(1, 5)),
                'updatedAt': now
            })
            variant_id_counter += 1

    variants_df = pd.DataFrame(variants_data)

    # --- SESUAIKAN TIPE DATA BIGINT(20) ---
    variants_df['id'] = variants_df['id'].astype(np.int64)
    variants_df['productId'] = variants_df['productId'].astype(np.int64)
    variants_df['finalPrice'] = variants_df['finalPrice'].astype(np.int64)

    return variants_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data products...")
try:
    products_df = pd.read_csv('products.csv')
except FileNotFoundError as e:
    print(f"Error: products.csv tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE PRODUCT VARIANTS ---
product_variants_df = generate_product_variants(products_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
product_variants_df.to_csv('product_variants.csv', index=False)
print(f"Data product_variants.csv berhasil dibuat. Total {len(product_variants_df)} varian dari {len(product_variants_df['productId'].unique())} produk.")
```

### 5. product_variant_options

```python
# --- FUNGSI GENERATE PRODUCT VARIANT OPTIONS ---

def generate_product_variant_options(product_variants_df, variant_options_df, variant_types_df, products_df):
    """
    Membuat relasi many-to-many antara product_variants dan variant_options.
    Sekaligus menghitung dan memperbarui finalPrice di product_variants_df.
    """
    now = datetime.now()
    variant_options_rel_data = []
    rel_id_counter = 1

    # Pre-calculated data structures
    option_type_map = variant_options_df.set_index('id')['variantTypeId'].to_dict()
    option_price_map = variant_options_df.set_index('id')['additionalPrice'].to_dict()

    # Inisialisasi dictionary untuk menyimpan update harga final
    price_updates = {}

    for _, variant in product_variants_df.iterrows():
        variant_id = variant['id']
        product_id = variant['productId']

        # Ambil base price dari produk induk
        base_price = products_df[products_df['id'] == product_id]['basePrice'].iloc[0]
        total_additional_price = 0

        # Cari semua variant types yang ada (GLOBAL dan STORE)
        available_type_ids = variant_types_df['id'].tolist()

        # Pilih 1 hingga 3 tipe varian acak yang akan dimasukkan ke varian ini
        num_types_to_use = random.randint(1, min(3, len(available_type_ids)))
        selected_type_ids = random.sample(available_type_ids, k=num_types_to_use)

        for type_id in selected_type_ids:
            # Filter opsi yang sesuai dengan variantTypeId yang dipilih
            relevant_options = variant_options_df[
                variant_options_df['variantTypeId'] == type_id
            ]['id'].tolist()

            if relevant_options:
                # Pilih satu opsi acak dari relevant options
                option_id = random.choice(relevant_options)

                # 1. Hitung harga tambahan
                additional_price = option_price_map.get(option_id, 0)
                total_additional_price += additional_price

                # 2. Buat relasi di tabel junction
                variant_options_rel_data.append({
                    'id': rel_id_counter,
                    'productVariantId': variant_id, # BIGINT(20)
                    'variantOptionId': option_id, # BIGINT(20)
                })
                rel_id_counter += 1

        # 3. Simpan update harga final (Base Price + Total Additional Price)
        price_updates[variant_id] = base_price + total_additional_price


    # --- PERBARUI FINAL PRICE DI product_variants_df ---

    # Buat salinan DataFrame untuk diperbarui
    updated_product_variants_df = product_variants_df.copy()

    # Lakukan update menggunakan map
    updated_product_variants_df['finalPrice'] = updated_product_variants_df['id'].map(price_updates).fillna(updated_product_variants_df['finalPrice'])
    updated_product_variants_df['finalPrice'] = updated_product_variants_df['finalPrice'].astype(np.int64)

    # --- FINAL DF GENERATION ---
    variant_options_rel_df = pd.DataFrame(variant_options_rel_data)

    # Sesuaikan tipe data BIGINT(20)
    variant_options_rel_df['id'] = variant_options_rel_df['id'].astype(np.int64)
    variant_options_rel_df['productVariantId'] = variant_options_rel_df['productVariantId'].astype(np.int64)
    variant_options_rel_df['variantOptionId'] = variant_options_rel_df['variantOptionId'].astype(np.int64)

    return variant_options_rel_df, updated_product_variants_df

# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data varian dan produk...")
try:
    products_df = pd.read_csv('products.csv')
    product_variants_df = pd.read_csv('product_variants.csv')
    variant_types_df = pd.read_csv('variant_types.csv')
    variant_options_df = pd.read_csv('variant_options.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV varian/produk tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE PRODUCT VARIANT OPTIONS ---
product_variant_options_df, updated_product_variants_df = generate_product_variant_options(
    product_variants_df, variant_options_df, variant_types_df, products_df
)

# --- SIMPAN DAN TAMPILKAN HASIL ---
# 1. Simpan tabel junction yang baru
product_variant_options_df.to_csv('product_variant_options.csv', index=False)
print(f"Data product_variant_options.csv berhasil dibuat. Total {len(product_variant_options_df)} relasi.")

# 2. Timpa/Update tabel product_variants dengan harga final yang benar
updated_product_variants_df.to_csv('product_variants.csv', index=False)
print("Data product_variants.csv berhasil diperbarui dengan finalPrice yang akurat.")
```

### 1. shifts

```python
# --- MAPPING SHIFT WAKTU NYATA ---
SHIFT_MAPPING = {
    'PAGI': {'start_hour_range': (7, 9), 'duration_range': (8, 9)},
    'SORE': {'start_hour_range': (15, 17), 'duration_range': (7, 8)},
    'MALAM': {'start_hour_range': (22, 0), 'duration_range': (7, 8)},
}

# --- FUNGSI GENERATE SHIFTS ---
def generate_shifts(store_users_df, num_shifts=30):
    """
    Membuat data dummy untuk tabel shifts, memilih secara acak dari shift Pagi, Sore, atau Malam.
    """
    now = datetime.now()
    shifts_data = []
    shift_id_counter = 1

    days_enum = ['MONDAY', 'TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY']
    active_user_store_ids = store_users_df[store_users_df['status'] == 'ACTIVE']['id'].tolist()

    if not active_user_store_ids:
        print("Peringatan: Tidak ada store_users yang aktif, tidak dapat membuat shift.")
        return pd.DataFrame(shifts_data)

    has_open_shift = False

    # Kumpulan shift yang harus dibuat (untuk memastikan keragaman)
    shift_types = list(SHIFT_MAPPING.keys()) * (num_shifts // 3)
    if num_shifts % 3 != 0:
        shift_types.extend(random.sample(list(SHIFT_MAPPING.keys()), k=num_shifts % 3))
    random.shuffle(shift_types)

    for i in range(num_shifts):
        user_store_id = random.choice(active_user_store_ids)
        shift_type = shift_types[i]
        shift_config = SHIFT_MAPPING[shift_type]

        # 1. Tentukan Waktu Mulai (Acak dalam 10 hari terakhir)
        if shift_type == 'MALAM':
            # Untuk malam (22.00-00.00), atur waktu mulai ke hari sebelumnya
            start_hour = random.randint(shift_config['start_hour_range'][0], 23) if shift_config['start_hour_range'][1] == 0 else random.randint(*shift_config['start_hour_range'])
            start_day_offset = random.randint(1, 10) # 1-10 hari yang lalu
            shift_start_base = now - timedelta(days=start_day_offset)
            shift_start_date = shift_start_base.replace(hour=start_hour, minute=random.randint(0, 59), second=0, microsecond=0)
        else:
            # Shift Pagi/Sore
            start_hour = random.randint(*shift_config['start_hour_range'])
            shift_start_base = now - timedelta(days=random.randint(0, 9))
            shift_start_date = shift_start_base.replace(hour=start_hour, minute=random.randint(0, 59), second=0, microsecond=0)

        # 2. Tentukan Durasi & Waktu Selesai
        duration_hours = random.randint(*shift_config['duration_range'])
        shift_end = shift_start_date + timedelta(hours=duration_hours)
        status = 'CLOSED'

        # 3. Logika Shift Terkini/Masa Depan
        if i == 0 and not has_open_shift:
            # Shift AKTIF/OPENED (Hanya 1)
            status = 'OPENED'
            shift_end = None
            has_open_shift = True
        elif random.random() < 0.15:
            # Shift PENDING (Jadwal masa depan)
            status = 'PENDING'
            shift_start_date = now + timedelta(hours=random.randint(1, 24))
            shift_end = None

        # Tentukan hari (sesuai ENUM)
        day_of_week = shift_start_date.weekday()
        day_name = days_enum[day_of_week]

        shifts_data.append({
            'id': shift_id_counter,
            'shiftStart': shift_start_date,
            'shiftEnd': shift_end,
            'days': day_name,
            'cashStatus': status,
            'userStoreId': user_store_id,
            'createdAt': shift_start_date,
            'updatedAt': now
        })
        shift_id_counter += 1

    shifts_df = pd.DataFrame(shifts_data)

    # Sesuaikan tipe data
    shifts_df['id'] = shifts_df['id'].astype(np.int64)
    shifts_df['userStoreId'] = shifts_df['userStoreId'].astype(np.int64)

    return shifts_df

# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data store_users...")
try:
    store_users_df = pd.read_csv('store_users.csv')
except FileNotFoundError as e:
    print(f"Error: store_users.csv tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE SHIFTS ---
shifts_df = generate_shifts(store_users_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
shifts_df.to_csv('shifts.csv', index=False)
print(f"Data shifts.csv berhasil dibuat. Total {len(shifts_df)} shift dengan pembagian waktu Pagi/Sore/Malam.")
```

### 2. cashier_cash

```python
# --- FUNGSI GENERATE CASHIER CASH ---
def generate_cashier_cash(shifts_df, stores_df, users_df, store_users_df):
    """
    Membuat data dummy untuk tabel cashier_cash, hanya untuk shift CLOSED dan OPENED.
    """
    now = datetime.now()
    cashier_cash_data = []
    cash_id_counter = 1

    # Filter shift yang relevan
    relevant_shifts_df = shifts_df[shifts_df['cashStatus'].isin(['CLOSED', 'OPENED'])].copy()

    # Gabungkan data untuk mendapatkan storeId dan userId
    shifts_with_store_user = relevant_shifts_df.merge(
        store_users_df[['id', 'userId', 'storeId']],
        left_on='userStoreId',
        right_on='id',
        suffixes=('_shift', '_su')
    )

    # Rentang kas awal (500.000 hingga 1.000.000, dibulatkan ke 50.000)
    initial_cash_range = [i * 50000 for i in range(10, 21)]

    for _, shift in shifts_with_store_user.iterrows():
        shift_id = shift['id_shift']
        store_id = shift['storeId']
        user_id = shift['userId']
        status = shift['cashStatus']

        initial_cash = random.choice(initial_cash_range)
        final_cash = None
        system_cash = None
        cash_difference = None
        is_balanced = None
        notes = None

        if status == 'CLOSED':
            # --- SKENARIO SHIFT SELESAI (CLOSED) ---

            # 1. Tentukan Uang Menurut Sistem (Simulasi Transaksi)
            system_cash_increase = random.randint(200000, 1500000)
            system_cash = initial_cash + system_cash_increase

            # 2. Tentukan Keseimbangan Kas
            if random.random() < 0.85:
                # 85% Shift Seimbang
                is_balanced = 1 # True
                cash_difference = 0
                final_cash = system_cash
                notes = random.choice([None, "Kasir melaporkan tidak ada kendala.", "Tutup kas sesuai data sistem."])
            else:
                # 15% Shift Tidak Seimbang (Kurang atau Lebih)
                is_balanced = 0 # False
                selisih = random.choice([
                    random.randint(-50000, -1000),
                    random.randint(1000, 50000)
                ])
                selisih = round(selisih, -2)

                cash_difference = selisih
                final_cash = system_cash + cash_difference

                notes = f"Terdapat selisih kas sebesar Rp{abs(selisih)}. Dicurigai karena kesalahan input kembalian."
                if cash_difference > 0:
                     notes = "Terdapat kelebihan kas. Dicurigai karena kasir salah memberikan kembalian."

            final_cash = final_cash if final_cash is not None else None

        elif status == 'OPENED':
            # --- SKENARIO SHIFT AKTIF (OPENED) ---
            notes = "Shift sedang berjalan."


        cashier_cash_data.append({
            'id': cash_id_counter,
            'shiftId': shift_id,
            'userId': user_id,
            'storeId': store_id,
            'initialCash': initial_cash,
            'finalCash': final_cash,
            'systemCash': system_cash,
            'cashDifference': cash_difference,
            'notes': notes,
            'isBalanced': is_balanced,
            'createdAt': shift['createdAt'],
            'updatedAt': now
        })
        cash_id_counter += 1

    cashier_cash_df = pd.DataFrame(cashier_cash_data)

    # Sesuaikan tipe data
    cashier_cash_df['id'] = cashier_cash_df['id'].astype(np.int64)
    cashier_cash_df['userId'] = cashier_cash_df['userId'].astype(np.int64)
    cashier_cash_df['finalCash'] = cashier_cash_df['finalCash'].astype('Int64')
    cashier_cash_df['systemCash'] = cashier_cash_df['systemCash'].astype('Int64')
    cashier_cash_df['cashDifference'] = cashier_cash_df['cashDifference'].astype('Int64')
    cashier_cash_df['isBalanced'] = cashier_cash_df['isBalanced'].astype('Int64')

    return cashier_cash_df

# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data shifts, stores, users, dan store_users...")
try:
    shifts_df = pd.read_csv('shifts.csv')
    stores_df = pd.read_csv('stores.csv')
    users_df = pd.read_csv('users.csv')
    store_users_df = pd.read_csv('store_users.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV shifts/store/user tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE CASHIER CASH ---
cashier_cash_df = generate_cashier_cash(shifts_df, stores_df, users_df, store_users_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
cashier_cash_df.to_csv('cashier_cash.csv', index=False)
print(f"Data cashier_cash.csv berhasil dibuat. Total {len(cashier_cash_df)} entitas kas.")
```

### 3. purchase_orders

```python
# --- FUNGSI GENERATE PURCHASE ORDERS ---

def generate_purchase_orders(suppliers_df, stores_df, num_orders=60):
    """
    Membuat data dummy untuk tabel purchase_orders.
    """
    now = datetime.now()
    purchase_orders_data = []
    order_id_counter = 1

    supplier_ids = suppliers_df['id'].tolist()
    store_ids = stores_df['id'].tolist()

    # Status ENUM: PENDING, SUCCESS, DELIVERED, CANCELED
    status_distribution = {
        'DELIVERED': 0.60,  # Mayoritas: Sudah diterima toko
        'SUCCESS': 0.20,    # Sudah dikirim/Diproses
        'PENDING': 0.10,    # Masih menunggu konfirmasi
        'CANCELED': 0.10    # Dibatalkan
    }

    for i in range(num_orders):
        supplier_id = random.choice(supplier_ids)
        store_id = random.choice(store_ids)

        # Tentukan orderDate acak dalam 6 bulan terakhir
        order_date = now - timedelta(days=random.randint(1, 180),
                                     hours=random.randint(1, 24),
                                     minutes=random.randint(1, 60))

        # Tentukan status
        rand_val = random.random()
        status = 'DELIVERED'
        if rand_val < status_distribution['CANCELED']:
            status = 'CANCELED'
        elif rand_val < status_distribution['CANCELED'] + status_distribution['PENDING']:
            status = 'PENDING'
        elif rand_val < status_distribution['CANCELED'] + status_distribution['PENDING'] + status_distribution['SUCCESS']:
            status = 'SUCCESS'

        # totalAmount akan diisi 0 dulu dan akan dihitung di Tahap 5.4
        total_amount = 0

        purchase_orders_data.append({
            'id': order_id_counter,
            'supplierId': supplier_id, # INT(11)
            'storeId': store_id,       # INT(11)
            'totalAmount': total_amount, # INT(11) (Update di Tahap 5.4)
            'status': status,          # ENUM
            'orderDate': order_date,   # DATETIME(3)
        })
        order_id_counter += 1

    purchase_orders_df = pd.DataFrame(purchase_orders_data)

    # Sesuaikan tipe data
    purchase_orders_df['id'] = purchase_orders_df['id'].astype(np.int64)
    purchase_orders_df['supplierId'] = purchase_orders_df['supplierId'].astype(np.int64)
    purchase_orders_df['storeId'] = purchase_orders_df['storeId'].astype(np.int64)

    return purchase_orders_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data suppliers dan stores...")
try:
    # Asumsi suppliers_df dan stores_df sudah dimuat.
    suppliers_df = pd.read_csv('suppliers.csv')
    stores_df = pd.read_csv('stores.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV suppliers/stores tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE PURCHASE ORDERS ---
purchase_orders_df = generate_purchase_orders(suppliers_df, stores_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
purchase_orders_df.to_csv('purchase_orders.csv', index=False)
print(f"Data purchase_orders.csv berhasil dibuat. Total {len(purchase_orders_df)} pesanan pembelian.")
```

### 4. purchase_order_items

```python
# --- FUNGSI GENERATE PURCHASE ORDER ITEMS ---

def generate_purchase_order_items(purchase_orders_df, products_df, product_categories_df):
    """
    Membuat detail item untuk setiap PO, menghitung totalAmount, dan menggunakan
    product_categories_df untuk logika expiredDate yang benar.
    """
    now = datetime.now()
    po_items_data = []
    item_id_counter = 1

    relevant_statuses = ['SUCCESS', 'DELIVERED', 'PENDING']
    relevant_pos_df = purchase_orders_df[
        purchase_orders_df['status'].isin(relevant_statuses)
    ].copy()

    all_product_ids = products_df['id'].tolist()
    product_map = products_df.set_index('id').to_dict('index')
    po_total_amounts = {}

    # --- LOGIKA KATEGORI KADALUARSA ---
    # 1. Buat Map Product ID -> Category ID
    product_id_to_category_id = products_df.set_index('id')['productCategoryId'].to_dict()
    # 2. Buat Map Category ID -> Category Name
    category_id_to_name = product_categories_df.set_index('id')['name'].to_dict()

    # Daftar Kategori yang memerlukan Tanggal Kedaluwarsa
    EXPIRE_DATE_CATEGORIES = [
        'Makanan Utama', 'Camilan dan Jajanan', 'Makanan Beku', 'Roti dan Kue', 'Es Krim',
        'Minuman Berbasis Kopi', 'Minuman Berbasis Teh', 'Minuman Berbasis Cokelat',
        'Minuman Kesehatan', 'Obat Resep & Obat Bebas', 'Suplemen dan Vitamin',
        'Alat Medis', 'Popok dan Produk Kebersihan', 'Makanan Bayi dan Susu Formula',
        'Perawatan Diri', 'Obat-obatan dan P3K', 'Makanan Hewan Peliharaan'
    ]

    for _, po in relevant_pos_df.iterrows():
        po_id = po['id']
        po_total = 0

        # Pilih Produk untuk PO ini (2 hingga 6 items)
        num_items = random.randint(2, 6)
        selected_product_ids = random.sample(all_product_ids, k=min(num_items, len(all_product_ids)))

        for product_id in selected_product_ids:
            product_data = product_map[product_id]

            # Mendapatkan Nama Kategori
            category_id = product_id_to_category_id[product_id]
            category_name = category_id_to_name.get(category_id)

            quantity = random.randint(20, 200)

            # Harga Modal (Capital Price): 20% - 40% di bawah Base Price (Harga Jual)
            discount_margin = random.uniform(0.20, 0.40)
            capital_price = int(product_data['basePrice'] * (1 - discount_margin))
            capital_price = round(capital_price, -3)

            # Logika Tanggal Kedaluwarsa
            expired_date = None
            if category_name in EXPIRE_DATE_CATEGORIES:
                # Kadaluarsa 3-18 bulan dari sekarang
                months_ahead = random.randint(3, 18)
                expired_date = now + relativedelta(months=months_ahead)

            subtotal = quantity * capital_price
            po_total += subtotal

            po_items_data.append({
                'id': item_id_counter,
                'purchaseOrderId': po_id,
                'productId': product_id,
                'quantity': quantity,
                'capitalPrice': capital_price,
                'expiredDate': expired_date,
            })
            item_id_counter += 1

        # Simpan Total Amount PO
        po_total_amounts[po_id] = po_total

    po_items_df = pd.DataFrame(po_items_data)
    po_items_df['productId'] = po_items_df['productId'].astype(np.int64)
    po_items_df['capitalPrice'] = po_items_df['capitalPrice'].astype(np.int64)

    return po_items_df, po_total_amounts

# --- EKSEKUSI ---

# 1. MEMBACA DATA
print("Memuat data purchase_orders, products, dan product_categories...")
try:
    purchase_orders_df = pd.read_csv('purchase_orders.csv')
    products_df = pd.read_csv('products.csv')
    product_categories_df = pd.read_csv('product_categories.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV tidak ditemukan: {e}")
    exit()

# 2. GENERATE ITEM DAN HITUNG TOTAL
po_items_df, po_totals = generate_purchase_order_items(purchase_orders_df, products_df, product_categories_df)

# 3. PERBARUI totalAmount DI purchase_orders_df
updated_po_df = purchase_orders_df.copy()
updated_po_df['totalAmount'] = updated_po_df['id'].map(po_totals).fillna(0)
updated_po_df['totalAmount'] = updated_po_df['totalAmount'].astype(np.int64)

# 4. SIMPAN DATA
po_items_df.to_csv('purchase_order_items.csv', index=False)
updated_po_df.to_csv('purchase_orders.csv', index=False)

print(f"\nData purchase_order_items.csv berhasil dibuat. Total {len(po_items_df)} item.")
print("Data purchase_orders.csv berhasil diperbarui dengan totalAmount yang benar.")
```

### 5. warehouse_stocks

```python
# --- FUNGSI GENERATE WAREHOUSE STOCKS ---

def generate_warehouse_stocks(purchase_orders_df, po_items_df):
    """
    Membuat stok produk di gudang berdasarkan Purchase Orders yang sudah DELIVERED.
    Setiap baris item PO yang DELIVERED akan menjadi satu baris stok.
    """
    now = datetime.now()
    stocks_data = []
    stock_id_counter = 1

    # 1. Filter PO yang statusnya DELIVERED
    delivered_pos_df = purchase_orders_df[
        purchase_orders_df['status'] == 'DELIVERED'
    ][['id', 'storeId', 'orderDate']].copy() # Ambil kolom yang relevan

    if delivered_pos_df.empty:
        print("Peringatan: Tidak ada Purchase Orders dengan status DELIVERED. Stok gudang akan kosong.")
        return pd.DataFrame(stocks_data)

    # 2. Gabungkan item PO dengan Store ID dari PO DELIVERED
    delivered_items_df = po_items_df.merge(
        delivered_pos_df,
        left_on='purchaseOrderId',
        right_on='id',
        suffixes=('_item', '_po')
    )

    # 3. Buat entri stok untuk setiap item yang diterima
    for _, item in delivered_items_df.iterrows():

        # Gunakan created_at dari PO untuk stok
        created_at_time = item['orderDate']

        stocks_data.append({
            'id': stock_id_counter,
            'productId': item['productId'], # BIGINT(20)
            'storeId': item['storeId'],     # INT(11)
            'purchaseOrderId': item['purchaseOrderId'], # INT(11) NULLABLE (di sini terisi)
            'stock': item['quantity'],      # INT(11)
            'createdAt': created_at_time,
            'updatedAt': now
        })
        stock_id_counter += 1

    stocks_df = pd.DataFrame(stocks_data)

    # Sesuaikan tipe data
    stocks_df['id'] = stocks_df['id'].astype(np.int64)
    stocks_df['productId'] = stocks_df['productId'].astype(np.int64)
    stocks_df['storeId'] = stocks_df['storeId'].astype(np.int64)
    stocks_df['purchaseOrderId'] = stocks_df['purchaseOrderId'].astype('Int64')

    return stocks_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data purchase_orders dan purchase_order_items...")
try:
    purchase_orders_df = pd.read_csv('purchase_orders.csv')
    po_items_df = pd.read_csv('purchase_order_items.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV purchase_orders/purchase_order_items tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE WAREHOUSE STOCKS ---
warehouse_stocks_df = generate_warehouse_stocks(purchase_orders_df, po_items_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
warehouse_stocks_df.to_csv('warehouse_stocks.csv', index=False)
print(f"Data warehouse_stocks.csv berhasil dibuat. Total {len(warehouse_stocks_df)} entitas stok.")
```

```python
# --- FUNGSI GENERATE CUSTOMERS---

def generate_customers(stores_df, store_users_df, generated_phone_numbers, num_customers=40):
    """
    Membuat data dummy untuk tabel customers, diikat ke store dan store_user aktif.
    """
    now = datetime.now()
    customers_data = []
    customer_id_counter = 1

    stores_list = stores_df['id'].tolist()

    # Filter store_users aktif yang akan mendaftarkan pelanggan
    active_store_users_df = store_users_df[store_users_df['status'] == 'ACTIVE'].copy()

    for i in range(num_customers):
        # 1. Pilih Toko Acak
        store_id = random.choice(stores_list)

        # 2. Filter Store Users yang Aktif di Toko Tersebut
        relevant_store_users = active_store_users_df[
            active_store_users_df['storeId'] == store_id
        ]['id'].tolist()

        if not relevant_store_users:
            # Jika tidak ada karyawan aktif di toko ini (seharusnya jarang terjadi), lewati
            continue

        # 3. Pilih Karyawan yang Mendaftarkan Pelanggan (storeUserId)
        store_user_id = random.choice(relevant_store_users)

        # 4. Generate Data Pelanggan
        customer_name = generate_gender_name()

        customers_data.append({
            'id': customer_id_counter,
            'storeId': store_id,            # INT(11)
            'storeUserId': store_user_id,   # BIGINT(20)
            'name': customer_name,          # Nama Realistis
            'phoneNumber': generate_phone(generated_phone_numbers), # Nomor Unik (62...)
            'status': random.choice(['ACTIVE', 'INACTIVE', 'ACTIVE']), # Mayoritas aktif
            'createdAt': now - timedelta(days=random.randint(1, 90)),
            'updatedAt': now
        })
        customer_id_counter += 1

    customers_df = pd.DataFrame(customers_data)

    # Sesuaikan tipe data
    customers_df['id'] = customers_df['id'].astype(np.int64)
    customers_df['storeId'] = customers_df['storeId'].astype(np.int64)
    customers_df['storeUserId'] = customers_df['storeUserId'].astype(np.int64)

    return customers_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data stores dan store_users...")
try:
    # Asumsi data dari Tahap 6 sudah dimuat di sesi
    stores_df = pd.read_csv('stores.csv')
    store_users_df = pd.read_csv('store_users.csv')
    users_df = pd.read_csv('users.csv')

    # Set Phone Numbers yang Sudah Terpakai
    generated_phone_numbers = set(users_df['phoneNumber'].tolist())

except FileNotFoundError as e:
    print(f"Error: File CSV stores/store_users tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE CUSTOMERS ---
customers_df = generate_customers(stores_df, store_users_df, generated_phone_numbers)

# --- SIMPAN DAN TAMPILKAN HASIL ---
customers_df.to_csv('customers.csv', index=False)
print(f"Data customers.csv berhasil dibuat. Total {len(customers_df)} pelanggan.")
```

### 2. transactions

```python
# --- MAPPING KATEGORI UNTUK POLA WAKTU KHUSUS ---
# Kita perlu tahu store_type_id toko tersebut (akan digabungkan di eksekusi)
WEEKEND_PEAK_TYPES = [2, 7, 8, 10, 11] # Fashion, Hobi, Elektronik, Otomotif, Hewan
FNB_ID = 1 # Makanan & Minuman
MINIMARKET_ID = 12 # Minimarket & Kelontong

def get_realistic_transaction_date(store_id, stores_df):
    """Menghasilkan transactionDate yang mengikuti pola bulanan, mingguan, dan harian."""
    now = datetime.now()

    # 1. Tentukan Pola Bulanan (Awal/Akhir Bulan)
    days_in_month = (now - now.replace(day=1)).days + 1
    # 70% terjadi pada 7 hari awal atau 5 hari akhir bulan
    if random.random() < 0.7:
        if random.random() < 0.6:
            # 60% dari 70% terjadi di awal bulan (hari 1-7)
            day_offset = random.randint(0, 7)
        else:
            # 40% dari 70% terjadi di akhir bulan (hari days_in_month-5 hingga days_in_month)
            day_offset = random.randint(days_in_month - 5, days_in_month)

        # Ambil tanggal acak dalam 30 hari terakhir, lalu paksakan ke awal/akhir bulan
        transaction_date = now - timedelta(days=random.randint(0, 30))
        try:
             transaction_date = transaction_date.replace(day=day_offset)
        except ValueError:
             # Fallback jika hari ke-31 tidak ada
             transaction_date = transaction_date.replace(day=days_in_month)
    else:
        # 30% sisanya terjadi di pertengahan bulan
        transaction_date = now - timedelta(days=random.randint(0, 30))

    # 2. Tentukan Pola Mingguan (Weekend Peak)
    store_type_id = stores_df[stores_df['id'] == store_id]['storeTypeId'].iloc[0]
    is_weekend_peak = store_type_id in WEEKEND_PEAK_TYPES or store_type_id == FNB_ID

    if is_weekend_peak and random.random() < 0.6:
        # 60% chance terjadi di Sabtu (5) atau Minggu (6)
        weekend_day = random.choice([5, 6])

        # Cek apakah tanggal yang terpilih adalah weekend (Sabtu=5, Minggu=6)
        while transaction_date.weekday() not in [5, 6]:
             transaction_date += timedelta(days=1)
             if (now - transaction_date).days > 30: # Batasi agar tidak terlalu jauh
                 transaction_date = now - timedelta(days=random.randint(0, 30))

    # 3. Tentukan Pola Harian (Jam Sibuk)

    # Default: Jam acak 9 pagi hingga 8 malam
    hour_range = (9, 20)

    if store_type_id == FNB_ID:
        # F&B/Makanan: Puncak di jam makan siang (11-14) atau malam (18-20)
        hour_range = random.choice([(11, 14), (18, 20)])
    elif store_type_id == MINIMARKET_ID:
        # Minimarket: Puncak sore/malam (16-22)
        hour_range = (16, 22)

    start_hour = random.randint(*hour_range)

    # Gabungkan tanggal dan waktu
    transaction_date = transaction_date.replace(hour=start_hour, minute=random.randint(0, 59), second=0, microsecond=0)

    return transaction_date

# --- FUNGSI GENERATE TRANSACTIONS ---

def generate_transactions(stores_df, customers_df, users_df, payment_methods_df, roles_df):
    """
    Membuat data dummy untuk tabel transactions.
    Diskon nominal transaksi (transactions.discount) dibatasi MAX Rp5.000.
    """
    now = datetime.now()
    transactions_data = []

    # Kumpulkan data kunci
    stores_list = stores_df['id'].tolist()
    customers_list = customers_df['id'].tolist()
    payment_methods_list = payment_methods_df['id'].tolist()

    # Logika Pencarian Kasir
    kasir_role_id = roles_df[roles_df['roleName'] == 'Kasir']['id'].iloc[0]
    kasir_users_list = users_df[users_df['roleId'] == kasir_role_id]['id'].tolist()

    if not kasir_users_list:
        raise ValueError("Tidak ada pengguna dengan peran Kasir ditemukan. Periksa users.csv dan roles.csv.")

    NUM_TRANSACTIONS = 100
    generated_invoice_ids = set()

    # Status ENUM
    status_distribution = {'PAID': 0.85, 'PENDING': 0.10, 'CANCELED': 0.05}

    for i in range(NUM_TRANSACTIONS):

        # 1. TANGGAL TRANSAKSI REALISTIS
        store_id = random.choice(stores_list)
        transaction_date = get_realistic_transaction_date(store_id, stores_df)

        # 2. Invoice ID Unik
        while True:
            invoice_id = f"INV-{transaction_date.strftime('%y%m%d')}-{i+1:04d}"
            if invoice_id not in generated_invoice_ids:
                generated_invoice_ids.add(invoice_id)
                break

        # 3. Pilih Store, Customer, Kasir, Payment Method acak
        customer_id = random.choice(customers_list)
        user_id_kasir = random.choice(kasir_users_list)
        payment_method_id = random.choice(payment_methods_list)

        # 4. Status Transaksi
        rand_val = random.random()
        status = 'PAID'
        if rand_val < status_distribution['CANCELED']:
            status = 'CANCELED'
        elif rand_val < status_distribution['CANCELED'] + status_distribution['PENDING']:
            status = 'PENDING'

        # 5. Diskon Transaksi (FINAL LOGIC: Diskon nominal, Maksimal Rp5.000)
        # 75% tanpa diskon (0), 25% diskon antara Rp1.000 hingga Rp5.000
        transaction_discount = random.choice([0, 0, 0, random.randint(1, 5) * 1000]) # <-- PERBAIKAN DI SINI
        transaction_discount = round(transaction_discount, -3)

        # totalAmount, cash, change akan dihitung di Tahap 6.3
        total_amount = 0; cash = 0; change = 0

        transactions_data.append({
            'id': i + 1, 'invoiceId': invoice_id, 'customerId': customer_id, 'userId': user_id_kasir,
            'storeId': store_id, 'paymentMethodId': payment_method_id, 'totalAmount': total_amount,
            'cash': cash, 'change': change, 'discount': transaction_discount,
            'transactionDate': transaction_date, 'status': status, 'createdAt': transaction_date,
            'updatedAt': now
        })

    transactions_df = pd.DataFrame(transactions_data)

    # Sesuaikan tipe data
    transactions_df['id'] = transactions_df['id'].astype(np.int64)
    transactions_df['userId'] = transactions_df['userId'].astype(np.int64)
    transactions_df['totalAmount'] = transactions_df['totalAmount'].astype(np.int64)
    transactions_df['cash'] = transactions_df['cash'].astype(np.int64)
    transactions_df['change'] = transactions_df['change'].astype(np.int64)
    transactions_df['discount'] = transactions_df['discount'].astype(np.int64)

    return transactions_df

# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN (FIX KEYERROR) ---
print("Memuat data stores, customers, users, payment_methods, dan roles...")
try:
    stores_df = pd.read_csv('stores.csv')
    customers_df = pd.read_csv('customers.csv')
    users_df = pd.read_csv('users.csv')
    payment_methods_df = pd.read_csv('payment_methods.csv')
    roles_df = pd.read_csv('roles.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE TRANSACTIONS ---
transactions_df = generate_transactions(stores_df, customers_df, users_df, payment_methods_df, roles_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
transactions_df.to_csv('transactions.csv', index=False)
print(f"Data transactions.csv berhasil dibuat. Total {len(transactions_df)} transaksi.")
```

### 3. transaction_details

```python
# --- FUNGSI GENERATE TRANSACTION DETAILS ---

def generate_transaction_details(transactions_df, products_df, product_variants_df, variant_options_df, payment_methods_df):
    """
    Membuat detail item untuk setiap transaksi dan menghitung/memperbarui
    totalAmount, cash, dan change di transactions_df.
    """
    now = datetime.now()
    details_data = []
    detail_id_counter = 1

    # Pre-calculated data structures
    product_map = products_df.set_index('id')[['basePrice', 'discount']].to_dict('index')
    variant_map = product_variants_df.set_index('id')[['productId', 'finalPrice']].to_dict('index')
    payment_map = payment_methods_df.set_index('id')['isCash'].to_dict()

    # Dictionaries untuk menyimpan update transaksi
    transaction_updates = {}

    for _, transaction in transactions_df.iterrows():
        transaction_id = transaction['id']
        transaction_status = transaction['status']
        trans_discount = transaction['discount']
        payment_method_id = transaction['paymentMethodId']
        is_cash = payment_map.get(payment_method_id, 'NO')

        # Hanya proses transaksi yang PAID atau PENDING
        if transaction_status in ['PAID', 'PENDING']:

            total_transaksi_sebelum_diskon = 0

            # Tentukan jumlah item per transaksi (2 hingga 8 item)
            num_items = random.randint(2, 8)

            # Pilih 20% transaksi yang akan memiliki produk bervarian
            has_variants = random.random() < 0.20

            # 1. Pilih Produk dan Variannya
            selected_product_ids = random.sample(products_df['id'].tolist(), k=min(num_items, len(products_df)))

            for product_id in selected_product_ids:

                # Cek apakah produk ini memiliki varian
                product_variants = product_variants_df[product_variants_df['productId'] == product_id]

                selected_variant_id = None
                selected_option_id = None

                # Tentukan Harga Satuan Awal (unitPrice)
                unit_price = product_map[product_id]['basePrice']

                # LOGIKA VARIAN
                if has_variants and not product_variants.empty:
                    # Jika ya, pilih satu varian
                    selected_variant = product_variants.sample(1).iloc[0]
                    selected_variant_id = selected_variant['id']

                    # Harga Satuan menjadi Final Price dari varian
                    unit_price = selected_variant['finalPrice']

                    # Cek opsi varian yang terikat untuk referensi
                    variant_options_rel = variant_options_df[
                        variant_options_df['variantTypeId'].isin(
                            product_variants_df[product_variants_df['id'] == selected_variant_id].merge(
                                variant_options_df, left_on='id', right_on='id', suffixes=('_pv', '_vo')
                            )['variantTypeId'].unique()
                        )
                    ]['id'].tolist()

                    if variant_options_rel:
                        selected_option_id = random.choice(variant_options_rel)

                # Kuantitas item (1 hingga 4)
                quantity = random.randint(1, 4)

                # Hitung subtotal
                subtotal = quantity * unit_price
                total_transaksi_sebelum_diskon += subtotal

                details_data.append({
                    'id': detail_id_counter,
                    'transactionId': transaction_id,      # BIGINT(20)
                    'productId': product_id,              # BIGINT(20)
                    'productVariantId': selected_variant_id, # BIGINT(20) NULLABLE
                    'quantity': quantity,                 # INT(11)
                    'unitPrice': unit_price,              # BIGINT(20)
                    'subtotal': subtotal,                 # BIGINT(20)
                    'createdAt': transaction['transactionDate'],
                    'updatedAt': now,
                    'variantOptionId': selected_option_id  # BIGINT(20) NULLABLE
                })
                detail_id_counter += 1

            # 2. Perhitungan Akhir Transaksi (totalAmount, cash, change)

            # Kurangi dengan diskon nominal transaksi
            final_total_amount = total_transaksi_sebelum_diskon - trans_discount

            # Jika total kurang dari 0 (misalnya diskon terlalu besar), set minimal 1000
            if final_total_amount <= 0:
                 final_total_amount = random.randint(1000, 5000)

            # Logika Pembayaran
            if is_cash == 'YES':
                # Pembayaran Tunai: Uang yang dibayarkan selalu lebih/sama dengan total
                cash_paid = final_total_amount + random.choice([0, 0, 0, random.randint(1000, 50000)])
                cash_paid = round(cash_paid, -3) # Bulatkan ke ribuan
                change_amount = cash_paid - final_total_amount
            else:
                # Pembayaran Non-Tunai: Cash = Total Amount, Change = 0
                cash_paid = final_total_amount
                change_amount = 0

            transaction_updates[transaction_id] = {
                'totalAmount': final_total_amount,
                'cash': cash_paid,
                'change': change_amount
            }

    details_df = pd.DataFrame(details_data)

    # --- SESUAIKAN TIPE DATA DETAIL ---
    details_df['transactionId'] = details_df['transactionId'].astype(np.int64)
    details_df['productId'] = details_df['productId'].astype(np.int64)
    details_df['productVariantId'] = details_df['productVariantId'].astype('Int64')
    details_df['unitPrice'] = details_df['unitPrice'].astype(np.int64)
    details_df['subtotal'] = details_df['subtotal'].astype(np.int64)
    details_df['variantOptionId'] = details_df['variantOptionId'].astype('Int64')

    return details_df, transaction_updates

# --- EKSEKUSI ---

# 1. MEMBACA DATA
print("Memuat data transaksi, produk, varian, opsi, dan pembayaran...")
try:
    transactions_df = pd.read_csv('transactions.csv')
    products_df = pd.read_csv('products.csv')
    product_variants_df = pd.read_csv('product_variants.csv')
    variant_options_df = pd.read_csv('variant_options.csv')
    payment_methods_df = pd.read_csv('payment_methods.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV tidak ditemukan: {e}")
    exit()

# 2. GENERATE ITEM DAN HITUNG TOTAL
details_df, transaction_updates = generate_transaction_details(
    transactions_df, products_df, product_variants_df, variant_options_df, payment_methods_df
)

# 3. PERBARUI transactions_df
updated_transactions_df = transactions_df.copy()
updated_transactions_df['totalAmount'] = updated_transactions_df['id'].map({k: v['totalAmount'] for k, v in transaction_updates.items()}).fillna(0)
updated_transactions_df['cash'] = updated_transactions_df['id'].map({k: v['cash'] for k, v in transaction_updates.items()}).fillna(0)
updated_transactions_df['change'] = updated_transactions_df['id'].map({k: v['change'] for k, v in transaction_updates.items()}).fillna(0)

# Perlu konversi ke Int64 karena mungkin ada NaN jika transaksi CANCELLED tidak memiliki detail
updated_transactions_df['totalAmount'] = updated_transactions_df['totalAmount'].astype(np.int64)
updated_transactions_df['cash'] = updated_transactions_df['cash'].astype(np.int64)
updated_transactions_df['change'] = updated_transactions_df['change'].astype(np.int64)

# 4. SIMPAN DATA
details_df.to_csv('transaction_details.csv', index=False)
updated_transactions_df.to_csv('transactions.csv', index=False)

print(f"\nData transaction_details.csv berhasil dibuat. Total {len(details_df)} detail.")
print("Data transactions.csv berhasil diperbarui dengan totalAmount, cash, dan change.")
```

```python
# --- FUNGSI GENERATE TAXES ---

def generate_taxes(transactions_df):
    """
    Membuat data dummy untuk tabel taxes.
    Menerapkan PPN 11% dan Pajak Daerah/Layanan 10% secara kondisional.
    """
    now = datetime.now()
    taxes_data = []
    tax_id_counter = 1

    # Filter transaksi yang relevan (PAID atau PENDING)
    relevant_transactions_df = transactions_df[
        transactions_df['status'].isin(['PAID', 'PENDING'])
    ].copy()

    for _, trans in relevant_transactions_df.iterrows():
        invoice_id = trans['invoiceId']

        # 1. Pajak Wajib: PPN (11%)
        taxes_data.append({
            'id': tax_id_counter,
            'transactionId': invoice_id,
            'name': 'PPN',
            'rate': 11, # 11%
        })
        tax_id_counter += 1

        # 2. Pajak Tambahan: Pajak Daerah/Layanan (10%) - Dikenakan 40% dari total transaksi
        if random.random() < 0.40:
            taxes_data.append({
                'id': tax_id_counter,
                'transactionId': invoice_id,
                'name': 'Pajak Layanan (PB01)',
                'rate': 10, # 10%
            })
            tax_id_counter += 1

    taxes_df = pd.DataFrame(taxes_data)

    # Sesuaikan tipe data
    taxes_df['id'] = taxes_df['id'].astype(np.int64)
    taxes_df['rate'] = taxes_df['rate'].astype(np.int64)

    return taxes_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data transactions...")
try:
    # transactions_df sudah diupdate dari Tahap 6.3
    transactions_df = pd.read_csv('transactions.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV transactions tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE TAXES ---
taxes_df = generate_taxes(transactions_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
taxes_df.to_csv('taxes.csv', index=False)
print(f"Data taxes.csv berhasil dibuat. Total {len(taxes_df)} entitas pajak.")
```

### 5. store_bills

```python
# --- FUNGSI GENERATE STORE BILLS ---

def generate_store_bills(stores_df, users_df, roles_df):
    """
    Membuat data dummy untuk tabel store_bills.
    """
    now = datetime.now()
    bills_data = []
    bill_id_counter = 1

    # Kumpulkan data kunci
    store_ids = stores_df['id'].tolist()

    # Filter Admin Toko dan Owner yang mungkin membayar tagihan
    admin_toko_id = roles_df[roles_df['roleName'] == 'Admin Toko']['id'].iloc[0]
    owner_id = roles_df[roles_df['roleName'] == 'Owner']['id'].iloc[0]

    # Users yang mungkin membayar (Admin Toko atau Owner)
    payer_user_ids = users_df[users_df['roleId'].isin([admin_toko_id, owner_id])]['id'].tolist()

    # Daftar Jenis Tagihan yang Realistis
    bill_types = {
        'Sewa Bulanan POS': (150000, 300000),  # Harga 150rb - 300rb
        'Tagihan Listrik': (100000, 500000),   # Harga 100rb - 500rb
        'Tagihan Air': (30000, 100000),        # Harga 30rb - 100rb
        'Internet & Telepon': (250000, 450000), # Harga 250rb - 450rb
    }

    for store_id in store_ids:
        # Buat 3 hingga 5 tagihan per toko
        num_bills = random.randint(3, 5)

        for i in range(num_bills):
            bill_name = random.choice(list(bill_types.keys()))
            price_range = bill_types[bill_name]
            price = random.randint(price_range[0], price_range[1])
            price = round(price, -3) # Bulatkan ke ribuan

            # Tentukan Tanggal Jatuh Tempo (3 bulan terakhir)
            on_date = now - timedelta(days=random.randint(1, 90))

            # Tentukan Status (80% PAID, 20% UNPAID)
            status = 'PAID' if random.random() < 0.80 else 'UNPAID'

            user_id_pembayar = None
            if status == 'PAID':
                # Jika PAID, tentukan siapa yang membayar (Owner atau Admin Toko)
                user_id_pembayar = random.choice(payer_user_ids)

            bills_data.append({
                'id': bill_id_counter,
                'storeId': store_id,        # INT(11)
                'name': bill_name,          # VARCHAR
                'status': status,           # ENUM
                'price': price,             # INT(11)
                'onDate': on_date,          # DATETIME(3)
                'createdAt': on_date,
                'updatedAt': now,
                'userId': user_id_pembayar, # BIGINT(20) NULLABLE
            })
            bill_id_counter += 1

    store_bills_df = pd.DataFrame(bills_data)

    # Sesuaikan tipe data
    store_bills_df['price'] = store_bills_df['price'].astype(np.int64)
    store_bills_df['userId'] = store_bills_df['userId'].astype('Int64') # BIGINT(20) NULLABLE

    return store_bills_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data stores, users, dan roles...")
try:
    stores_df = pd.read_csv('stores.csv')
    users_df = pd.read_csv('users.csv')
    roles_df = pd.read_csv('roles.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV stores/users/roles tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE STORE BILLS ---
store_bills_df = generate_store_bills(stores_df, users_df, roles_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
store_bills_df.to_csv('store_bills.csv', index=False)
print(f"Data store_bills.csv berhasil dibuat. Total {len(store_bills_df)} tagihan toko.")
```

### 6. user_activity

```python
# --- FUNGSI GENERATE USER ACTIVITY ---

def generate_user_activity(store_users_df, users_df, num_activities=180):
    """
    Membuat data dummy untuk tabel user_activity, mencatat aksi Admin Toko dan Kasir.
    """
    now = datetime.now()
    activity_data = []
    activity_id_counter = 1

    # 1. Filter User yang Relevan (Admin Toko & Kasir)
    # Kita hanya menggunakan user yang terdaftar di store_users
    active_store_user_data = store_users_df[store_users_df['status'] == 'ACTIVE'].merge(
        users_df[['id', 'roleId']], left_on='userId', right_on='id', suffixes=('_su', '_u')
    )

    # Kumpulkan daftar ID user dan store yang relevan untuk logging
    valid_activity_points = active_store_user_data[['userId', 'storeId']].drop_duplicates().values.tolist()

    # Daftar Jenis Aktivitas Realistis
    log_actions = [
        'Login berhasil',
        'Logout',
        'Membuka sesi kasir',
        'Mengubah harga produk',
        'Menambah stok baru',
        'Menerima pesanan pembelian (PO)',
        'Mencetak ulang struk transaksi',
        'Melihat laporan penjualan bulanan',
        'Melakukan penyesuaian stok (stok opname)',
        'Mendaftarkan pelanggan baru',
        'Mengubah data pelanggan',
        'Mengunci data shift (shift closure)'
    ]

    for i in range(num_activities):
        # 2. Pilih User dan Toko (Pasangan yang Valid)
        user_id, store_id = random.choice(valid_activity_points)

        # Tentukan Waktu Aktivitas (30 hari terakhir)
        action_time = now - timedelta(days=random.randint(0, 30),
                                      hours=random.randint(1, 24),
                                      minutes=random.randint(1, 60))

        # 3. Tentukan Aksi
        action = random.choice(log_actions)

        activity_data.append({
            'id': activity_id_counter,
            'userId': user_id,       # BIGINT(20)
            'storeId': store_id,     # INT(11)
            'action': action,        # VARCHAR
            'actionTime': action_time,
        })
        activity_id_counter += 1

    activity_df = pd.DataFrame(activity_data)

    # Sesuaikan tipe data
    activity_df['id'] = activity_df['id'].astype(np.int64)
    activity_df['userId'] = activity_df['userId'].astype(np.int64)
    activity_df['storeId'] = activity_df['storeId'].astype(np.int64)

    return activity_df


# --- MEMBACA KEMBALI DATA DARI FILE CSV YANG DIBUTUHKAN ---
print("Memuat data store_users dan users...")
try:
    # Data dari Tahap 6 dan sebelumnya
    store_users_df = pd.read_csv('store_users.csv')
    users_df = pd.read_csv('users.csv')
except FileNotFoundError as e:
    print(f"Error: File CSV store_users/users tidak ditemukan: {e}")
    exit()

# --- EKSEKUSI FUNGSI GENERATE USER ACTIVITY ---
user_activity_df = generate_user_activity(store_users_df, users_df)

# --- SIMPAN DAN TAMPILKAN HASIL ---
user_activity_df.to_csv('user_activity.csv', index=False)
print(f"Data user_activity.csv berhasil dibuat. Total {len(user_activity_df)} log aktivitas.")
```
